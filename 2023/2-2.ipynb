```python
game_dict = {}

for x in data.split('\n'):
    row = x.split(': ')
    game = int(row[0].split(' ')[1])
    grabs = row[1].split("; ")
    
    game_grabs = []
    for group in grabs:
        group_dict = {}
        for z in group.split(", "):
            key, value = z.split(" ")
            group_dict[value] = int(key)
        game_grabs.append(group_dict)
    
    game_dict[game] = game_grabs

game_dict
```

```python
answer = 0
for game_id, game in game_dict.items():
    min_possible = {}
    for hand in game:
        for color, number in hand.items():
            if not color in min_possible.keys():
                min_possible[color] = number
            elif min_possible[color] < number:
                min_possible[color] = number
    answer += np.prod(list(min_possible.values()))
print(answer)
        
```
