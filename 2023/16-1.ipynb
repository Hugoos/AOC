```python
from helpers import npNicePrint
import numpy as np
import ast

#helpers.npNicePrint(arr)


data = r""".|...\....
|.-.\.....
.....|-...
........|.
..........
.........\
..../.\\..
.-.-/..|..
.|....-|.\
..//.|...."""


with open("input_16.txt", 'r', encoding='raw_unicode_escape') as f:
    data = f.read().strip()

grid = {(x,y): val for y, r in enumerate(data.split('\n')) for x, val in enumerate(r)}
print(len(grid))
grid
```

```python
directions = {"N": np.array((0,-1)), "S": np.array((0, 1)), "W": np.array((-1, 0)), "E": np.array((1,0))}

movements = {"|": {"S": "S", "N": "N", "E": ["N", "S"], "W": ["N", "S"]},
             "-": {"E": "E", "W": "W", "N": ["W", "E"], "S": ["W", "E"]},
             "\\": {"S": "E", "W": "N", "E": "S", "N": "W"},
             "/": {"S": "W", "E": "N", "N": "E", "W": "S"},
             ".": {"S": "S", "E": "E", "W": "W", "N": "N"}
            }
```

```python
from collections import deque

start_coord = ((0,0), "S")
coords = deque([start_coord])
seen_coords = set()
while not len(coords) == 0:
    next_coord = coords.popleft()
    seen_coords.add(next_coord)
    coord, direction = next_coord

    coord = tuple(coord + directions[direction])
    
    if coord not in grid.keys():
        print(coord)
        print("invalid coordinate")
        continue
    tile = grid[tuple(coord)]
    
    directions_ = movements[tile].get(direction)
    for direction in directions_:
        candidate_coord = (coord, direction)
        #print(candidate_coord, seen_coords)
        if not candidate_coord in seen_coords:
            coords.append((coord, direction))
seen_coords
```
