```python
def get_coordinates(data, character):
    rows = data.split('\n')
    first_row = rows[0]
    last_row = rows[-1]

    # Find coordinates of "." in the first row
    first_row_coordinates = [(j, 0) for j, char in enumerate(first_row) if char == character]

    # Find coordinates of "." in the last row
    last_row_coordinates = [(j, len(rows) - 1) for j, char in enumerate(last_row) if char == character]

    return first_row_coordinates[0], last_row_coordinates[0]

# Get coordinates of "." in the first and last row
start_coord, end_coord = get_coordinates(data, '.')

# Print the results
print("Coordinates of '.' in the first row:", start_coord)
print("Coordinates of '.' in the last row:", end_coord)
```

```python
def create_graph(locations):
    graph = {}
    forced_movements = {"v": (0, 1), ">": (1, 0), "<": (-1,0), "^": (0,-1)}
    for current_location, value in locations.items():
        x, y = current_location
        if value == "#":
            continue
        if value in forced_movements.keys():
            neighs = [forced_movements[value]]
        else:
            neighs = [(-1, 0), (1, 0), (0, -1), (0, 1)]
            
        # Initialize the neighbors dictionary for the current vertex
        if current_location not in graph:
            graph[current_location] = {}
        
        # Add neighbors to the west, east, north, and south
        for i, j in neighs:
            neighbor_location = (x + i, y + j)

            # Check if the neighbor is within the bounds of the given locations
            if neighbor_location in locations:
                neighbor_value = locations[neighbor_location]

                # Add the neighbor to the current vertex with the corresponding value as weight
                if not neighbor_value == "#":
                    graph[current_location][neighbor_location] = 1

    return graph

graph = create_graph(grid)


# Print the resulting graph
for vertex, neighbors in graph.items():
    print(f"{vertex}: {neighbors}")
```

```python
%matplotlib notebook
import networkx as nx
import matplotlib.pyplot as plt
from collections import deque

forced_movements = {"v": (0, 1), ">": (1, 0), "<": (-1,0), "^": (0,-1)}

# Create a deque
nodes = deque([(start_coord, (-100,-100))])
#print(nodes)

G = nx.DiGraph()

while not len(nodes) == 0:
    node, came_from = nodes.popleft()
    #print(node, came_from)
    G.add_node(node)
    #print(node)
    x,y = node
    value = grid[node]
    
    if value in forced_movements.keys():
        neighs = [forced_movements[value]]
    else:
        neighs = [(-1, 0), (1, 0), (0, -1), (0, 1)]
    
    for i, j in neighs:
        neighbor_location = (x + i, y + j)
        if neighbor_location in grid and not neighbor_location == came_from:
            if not grid[neighbor_location] == "#":
                if not (((i,j) == (-1,0) and grid[neighbor_location] == ">") or ((i,j) == (0,-1) and grid[neighbor_location] == "v")): 
                    G.add_edge(node, neighbor_location)
                    nodes.append((neighbor_location, node))
                
for layer, nodes in enumerate(nx.topological_generations(G)):
    # `multipartite_layout` expects the layer as a node attribute, so add the
    # numeric layer value as a node attribute
    for node in nodes:
        G.nodes[node]["layer"] = layer
    
# Visualize the graph (optional)
#pos = nx.multipartite_layout(G, subset_key="layer")


# Draw the DAG
#nx.draw(G, pos, with_labels=True, font_weight='bold', node_size=700, node_color='skyblue', font_color='black', edge_color='gray', width=2, arrowsize=20)

# Display the graph
#plt.show()
```
