```python
def shift_ones_up(arr):
    rows, cols = arr.shape

    # Find positions of all 1's
    ones_positions = np.argwhere(arr == 1)

    # Iterate through each position of 1
    for pos in ones_positions:
        row, col = pos
        # Check if the position above is within bounds and is not blocked by a 2
        if row > 0 and arr[row - 1, col] == 0:
            # Swap values
            arr[row, col], arr[row - 1, col] = arr[row - 1, col], arr[row, col]

    return arr


grid_old = np.array([])

while not np.array_equal(grid_old, grid):
    grid_old = grid.copy()
    grid = shift_ones_up(grid)

npNicePrint(grid, convertDict={value: key for key, value in transform.items()})

ans = 0
rows, columns = grid.shape
for r,c in np.argwhere(grid == 1):
    ans += rows-r
ans
```
