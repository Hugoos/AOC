```python
from tqdm import tqdm
def shift_ones_up(arr):
    rows, cols = arr.shape

    # Find positions of all 1's
    ones_positions = np.argwhere(arr == 1)

    # Iterate through each position of 1
    for pos in ones_positions:
        row, col = pos
        # Check if the position above is within bounds and is not blocked by a 2
        if row > 0 and arr[row - 1, col] == 0:
            # Swap values
            arr[row, col], arr[row - 1, col] = arr[row - 1, col], arr[row, col]

    return arr



def shift_ones_west(arr):
    rows, cols = arr.shape

    # Find positions of all 1's
    ones_positions = np.argwhere(arr == 1)

    # Iterate through each position of 1
    for pos in ones_positions:
        row, col = pos
        # Check if the position to the left is within bounds and is not blocked by a 2
        if col > 0 and arr[row, col - 1] == 0:
            # Swap values
            arr[row, col], arr[row, col - 1] = arr[row, col - 1], arr[row, col]

    return arr

def shift_ones_east(arr):
    rows, cols = arr.shape

    # Find positions of all 1's
    ones_positions = np.argwhere(arr == 1)

    # Iterate through each position of 1
    for pos in ones_positions:
        row, col = pos
        # Check if the position to the right is within bounds and is not blocked by a 2
        if col < cols - 1 and arr[row, col + 1] == 0:
            # Swap values
            arr[row, col], arr[row, col + 1] = arr[row, col + 1], arr[row, col]

    return arr

def shift_ones_south(arr):
    rows, cols = arr.shape

    # Find positions of all 1's
    ones_positions = np.argwhere(arr == 1)

    # Iterate through each position of 1
    for pos in ones_positions:
        row, col = pos
        # Check if the position below is within bounds and is not blocked by a 2
        if row < rows - 1 and arr[row + 1, col] == 0:
            # Swap values
            arr[row, col], arr[row + 1, col] = arr[row + 1, col], arr[row, col]

    return arr
grid = grid_b.copy()
loads = []
for x in tqdm(range(1000)):
    load = 0
    grid_old = np.array([])
    while not np.array_equal(grid_old, grid):
        grid_old = grid.copy()
        grid = shift_ones_up(grid)

    #npNicePrint(grid, convertDict={value: key for key, value in transform.items()})
    
    grid_old = np.array([])
    while not np.array_equal(grid_old, grid):
        grid_old = grid.copy()
        grid = shift_ones_west(grid)

    #npNicePrint(grid, convertDict={value: key for key, value in transform.items()})
    
    grid_old = np.array([])
    while not np.array_equal(grid_old, grid):
        grid_old = grid.copy()
        grid = shift_ones_south(grid)

    #npNicePrint(grid, convertDict={value: key for key, value in transform.items()})
    
    grid_old = np.array([])
    while not np.array_equal(grid_old, grid):
        grid_old = grid.copy()
        grid = shift_ones_east(grid)

    #npNicePrint(grid, convertDict={value: key for key, value in transform.items()})

    rows, columns = grid.shape
    for r,c in np.argwhere(grid == 1):
        load += rows-r
    loads.append(load)
```
