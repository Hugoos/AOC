```python
import re

def is_symbol(input_str):
    # Define a regular expression pattern for symbols (excluding '.')
    pattern = re.compile(r'^[^\d\w.]$')

    # Use the pattern to match the input string
    match = pattern.match(input_str)

    # If there is a match, it's a symbol; otherwise, it's not
    return bool(match)
```

```python
def get_surrounding_coordinates(coord):
    x, y = coord
    surrounding_coordinates = []

    for i in range(-1, 2):
        for j in range(-1, 2):
            if i == 0 and j == 0:
                continue  # Skip the center coordinate (x, y)
            
            surrounding_coordinates.append((x + i, y + j))

    return surrounding_coordinates

center_coordinate = (3, 4)
surrounding_coords = get_surrounding_coordinates(center_coordinate)

print(f"Surrounding coordinates for {center_coordinate}: {surrounding_coords}")
```

```python
recording = False
number = ""
solution = 0
for y, row in enumerate(data_o.split('\n')): 
    for x, val in enumerate(row):
        if data[(x,y)].isdigit():
            number += data[(x,y)]
            for coord in get_surrounding_coordinates((x,y)):
                if coord in data:
                    if is_symbol(data[coord]):
                        recording = True
        elif recording:
            solution += int(number)
            number = ""
            recording = False
        else:
            number = ""
            
solution       
    
```

```python
game_dict = {}

for x in data.split('\n'):
    row = x.split(': ')
    game = int(row[0].split(' ')[1])
    grabs = row[1].split("; ")
    
    game_grabs = []
    for group in grabs:
        group_dict = {}
        for z in group.split(", "):
            key, value = z.split(" ")
            group_dict[value] = int(key)
        game_grabs.append(group_dict)
    
    game_dict[game] = game_grabs

game_dict
```

```python
target = {"red": 12, "green": 13, "blue": 14}
answer = 0
for game_id, game in game_dict.items():
    add = game_id
    for hand in game:
        for color, number in hand.items():
            if color in target.keys():
                if number > target[color]:
                    add = 0
    answer += add
print(answer)
        
```
