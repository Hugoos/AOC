```python
from helpers import npNicePrint
import numpy as np
import ast

#helpers.npNicePrint(arr)


data = """px{a<2006:qkq,m>2090:A,rfg}
pv{a>1716:R,A}
lnx{m>1548:A,A}
rfg{s<537:gd,x>2440:R,A}
qs{s>3448:A,lnx}
qkq{x<1416:A,crn}
crn{x>2662:A,R}
in{s<1351:px,qqz}
qqz{s>2770:qs,m<1801:hdj,R}
gd{a>3333:R,R}
hdj{m>838:A,pv}

{x=787,m=2655,a=1222,s=2876}
{x=1679,m=44,a=2067,s=496}
{x=2036,m=264,a=79,s=2244}
{x=2461,m=1339,a=466,s=291}
{x=2127,m=1623,a=2188,s=1013}"""



with open("input_19.txt", 'r') as f:
    data = f.read().strip()

rules_d, parts_d = data.split("\n\n")

rules = {}

for rule in rules_d.split("\n"):
    rule_name = rule.split("{")[0]
    rules[rule_name] = []
    rule_statements = rule.split("{")[1].split("}")[0]
    for rule_statement in rule_statements.split(","):
        rule_axioms = rule_statement.split(":")
        if len(rule_axioms) == 1:
            rules[rule_name].append(["1==1", rule_axioms[0]])
        else:
            rules[rule_name].append(rule_axioms)
            
parts = []
            
for part in parts_d.replace("=", ":").split("\n"):
    print(part)
    fixed_string = part.replace(':', '":').replace(',', ',"')
    fixed_string = fixed_string.replace('{', '{"')
    parts.append(ast.literal_eval(fixed_string))
    
rules, parts
```

```python
expression_paths = []
def rules_explorer(location,expression_queue):
    if location == "A":
        expression_paths.append(expression_queue)
        return
    if location == "R":
        return
    for expr, target in rules[location]:
        if not expr == "1==1":
            n_expression_queue = expression_queue.copy()
            n_expression_queue.append(expr)
            rules_explorer(target, n_expression_queue)
            if ">" in expr:
                expr = expr.replace(">", "<=")
            elif "<" in expr:
                expr = expr.replace("<", ">=")
            expression_queue.append(expr)
        elif expr == "1==1":
            rules_explorer(target, expression_queue)
        
        
expression_paths = []    
rules_explorer("in",[])       
expression_paths
```

```python
ranges_all = []
for exp in expression_paths:
    ranges = {
        "x": [1,4000],
        "m": [1,4000],
        "a": [1,4000],
        "s": [1,4000]
    }
    for rule in exp:
        target = rule[0]
        if "<=" in rule:
            ranges[target][1] = min(ranges[target][1], int(rule.split("<=")[1]))
        elif ">=" in rule:
            ranges[target][0] = max(ranges[target][0], int(rule.split(">=")[1]))
        elif "<" in rule:
            ranges[target][1] = min(ranges[target][1], int(rule.split("<")[1])-1)
        elif ">" in rule:
            ranges[target][0] = max(ranges[target][0], int(rule.split(">")[1])+1)
    ranges_all.append(ranges)
ranges_all
```

```python
ans = 0
for thing in ranges_all:
    thing_ans = 1
    for x in thing.values():
        thing_ans *= (x[1] - x[0]+1)
    ans += thing_ans
ans
```
