```python
from helpers import npNicePrint
import numpy as np
import ast

#helpers.npNicePrint(arr)


data = """19, 13, 30 @ -2,  1, -2
18, 19, 22 @ -1, -1, -2
20, 25, 34 @ -2, -2, -4
12, 31, 28 @ -1, -2, -1
20, 19, 15 @  1, -5, -3"""

with open("input_24.txt", 'r') as f:
    data = f.read().strip()

hail = []

for row in data.split("\n"):
    coords, velocities = row.split(" @ ")
    coords = [int(x) for x in coords.split(", ")]
    velocities = [int(x) for x in velocities.split(", ")]
    hail.append([coords, velocities])
hail
```

```python
def linear_function(x1, y1, vx, vy):
    x2 = x1 + vx
    y2 = y1 + vy
    # Calculate the slope (m)
    slope = (y2 - y1) / (x2 - x1)

    # Calculate the y-intercept (b)
    intercept = y1 - slope * x1

    # Define the linear function using a lambda function
    #linear_func = lambda x: slope * x + intercept
    
    return slope, intercept

for row in hail:
    row.append(linear_function(*row[0][:2], *row[1][:2]))
    
hail
```
