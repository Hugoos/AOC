```python
import helpers
import numpy as np

#helpers.npNicePrint(arr)

data = """....#........
.........#...
#............
.............
.............
........#....
.#...........
............#
.............
.............
.........#...
#....#......."""
with open("input_11.txt", 'r') as f:
    data = f.read().strip()

grid = {(x,y): val for y, r in enumerate(data.split('\n')) for x, val in enumerate(r) if val == "#"}

#start_coord = np.array([k for k,v in grid.items() if v == "S"][0])
grid
```

```python
y_max = len(data.split("\n"))
x_max = len(data.split("\n")[0])

x_coords = set([x[0] for x in grid.keys()])
y_coords = set([x[1] for x in grid.keys()])

x_coords_missing = set([x for x in range(x_max)]) - x_coords
y_coords_missing = set([y for y in range(y_max)]) - y_coords
x_coords_missing, y_coords_missing
```

```python
def num_smaller(num, missing):
    return len([x for x in missing if x < num])
```

```python
expanded_grid = {(k[0] + num_smaller(k[0], x_coords_missing), k[1] + num_smaller(k[1], y_coords_missing)): v for k,v in grid.items()}
expanded_grid
```

```python
def manhattan_distance(point1, point2):
    """
    Calculate the Manhattan distance between two points.

    Parameters:
    - point1, point2: Tuple or list representing the coordinates of the two points.

    Returns:
    - Manhattan distance between the two points.
    """
    return sum(abs(x - y) for x, y in zip(point1, point2))

# Example usage:
point_a = (1, 2)
point_b = (4, 6)

distance = manhattan_distance(point_a, point_b)
distance
```

```python
from itertools import combinations

def unique_combinations(elements):
    """
    Generate unique combinations of elements, excluding reversed duplicates.

    Parameters:
    - elements: List of elements.

    Returns:
    - List of unique combinations.
    """
    result = []
    seen_pairs = set()

    for combo in combinations(elements, 2):
        pair = tuple(sorted(combo))  # Sort the elements in each pair
        if pair not in seen_pairs:
            result.append(combo)
            seen_pairs.add(pair)

    return result

# Example usage:
my_list = [1, 2, 3, 4]

sum([manhattan_distance(combs[0], combs[1]) for combs in unique_combinations(expanded_grid.keys())])
```
