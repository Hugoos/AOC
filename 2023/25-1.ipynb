```python
from helpers import npNicePrint
import numpy as np
import ast

#helpers.npNicePrint(arr)


data = """jqt: rhn xhk nvd
rsh: frs pzl lsr
xhk: hfx
cmg: qnr nvd lhk bvb
rhn: xhk bvb hfx
bvb: xhk hfx
pzl: lsr hfx nvd
qnr: nvd
ntq: jqt hfx bvb xhk
nvd: lhk
lsr: lhk
rzs: qnr cmg lsr rsh
frs: qnr lhk lsr"""

with open("input_25.txt", 'r') as f:
    data = f.read().strip()

```

```python
%matplotlib notebook
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

for row in data.split("\n"):
    node, children = row.split(": ")
    G.add_node(node)
    for child in children.split(" "):
        G.add_edge(node, child)
        
#pos = nx.kamada_kawai_layout(G)
pos = nx.spring_layout(G)

# Draw the DAG
nx.draw(G, pos, with_labels=True, font_weight='bold', node_size=700, node_color='skyblue', font_color='black', edge_color='gray', width=2, arrowsize=20)

# Display the graph
plt.show()
```

```python
%matplotlib notebook
G.remove_edges_from([("mbq", "vgk"), ("thl", "nmv"), ("fzb", "fxr")])
        
#pos = nx.kamada_kawai_layout(G)
pos = nx.spring_layout(G)

# Draw the DAG
nx.draw(G, pos, with_labels=True, font_weight='bold', node_size=700, node_color='skyblue', font_color='black', edge_color='gray', width=2, arrowsize=20)

# Display the graph
plt.show()
```
