```python
def create_graph(locations):
    graph = {}

    for current_location, value in locations.items():
        x, y = current_location

        # Initialize the neighbors dictionary for the current vertex
        if current_location not in graph:
            graph[current_location] = {}

        # Add neighbors to the west, east, north, and south
        for i, j in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
            neighbor_location = (x + i, y + j)

            # Check if the neighbor is within the bounds of the given locations
            if neighbor_location in locations:
                neighbor_value = locations[neighbor_location]

                # Add the neighbor to the current vertex with the corresponding value as weight
                graph[current_location][neighbor_location] = neighbor_value

    return graph


graph = create_graph(grid)

# Print the resulting graph
for vertex, neighbors in graph.items():
    print(f"{vertex}: {neighbors}")
```

```python
import heapq

def dijkstra(graph, start):
    priority_queue = [(0, 0, start, (0, 0))]
    distances = {(vertex, x, y): float('infinity') for vertex in graph for x in range(11) for y in [(0,1), (0,-1), (1,0), (-1,0)]}
    distances[(start, 0, (0,0))] = 0

    while priority_queue:
        current_distance, num_forwards, current_vertex, heading = heapq.heappop(priority_queue)
        #print((current_vertex, num_forwards), heading)
        if current_distance > distances[(current_vertex, num_forwards, heading)]:
            continue
        
        for neighbor in graph[current_vertex]:
            x, y = neighbor
            value = graph[current_vertex][neighbor]
            new_heading = np.array(neighbor) - np.array(current_vertex)
            forward = np.array_equal(new_heading, heading)
            #if np.array_equal(heading,(0,0)): #first 
            if heading == (0,0):
                new_num_forwards = 1
            elif forward:
                new_num_forwards = num_forwards + 1
            else:
                new_num_forwards = 1
            # Check if moving to the neighbor violates the constraints
            
            
            #if forward and new_num_forwards < 10 and not np.array_equal(new_heading, tuple(np.multiply(heading, -1)))
            
            #if 4 <= new_num_forwards < 10 and not np.array_equal(new_heading, tuple(np.multiply(heading, -1))):
            #if not forward:
            #    print(current_vertex, neighbor, num_forwards)
            if (not forward and 4 <= num_forwards <= 10 and not np.array_equal(new_heading, tuple(np.multiply(heading, -1)))) or ((forward or heading == (0,0)) and new_num_forwards <= 10 and not np.array_equal(new_heading, tuple(np.multiply(heading, -1)))):
                #print(heading, new_heading)
                distance = current_distance + value
                #print(1, current_vertex, neighbor)
                if distance < distances[(neighbor, new_num_forwards, tuple(new_heading))]:
                    distances[(neighbor, new_num_forwards, tuple(new_heading))] = distance
                    #print(2, current_vertex, neighbor)
                    heapq.heappush(priority_queue, (distance, new_num_forwards, neighbor, tuple(new_heading)))
    return distances


start_vertex = (0, 0)
shortest_distances = dijkstra(graph, start_vertex)

#print(f"Shortest distances from {start_vertex}: {shortest_distances}")

min_distances = {vertex: float('infinity') for vertex in graph}
#max_distances = {vertex: 0 for vertex in graph}

for (node, distance, heading), val in shortest_distances.items():
    if not val > 999999999999999999999999 and distance >= 4:
        min_distances[node] = min(val, min_distances[node])
    #if np.array_equal(node, end_coord):
    #    print(val)
min_distances   
min_distances[end_coord]
```
