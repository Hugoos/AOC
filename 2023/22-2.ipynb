```python
from helpers import npNicePrint
import numpy as np
import ast

#helpers.npNicePrint(arr)


data = """1,0,1~1,2,1
0,0,2~2,0,2
0,2,3~2,2,3
0,0,4~0,2,4
2,0,5~2,2,5
0,1,6~2,1,6
1,1,8~1,1,9"""

with open("input_22.txt", 'r') as f:
    data = f.read().strip()

bricks = []

for coords in data.split("\n"):
    b_start, b_end = coords.split("~")
    b_start = [int(x) for x in b_start.split(",")]
    b_end = [int(x) for x in b_end.split(",")]
    bricks.append([b_start, b_end])
bricks
```

```python
def overlap(cubeA, cubeB):
    return (cubeA[1][0] >= cubeB[0][0]) and (cubeA[0][0] <= cubeB[1][0]) and \
           (cubeA[1][1] >= cubeB[0][1]) and (cubeA[0][1] <= cubeB[1][1]) and \
           (cubeA[1][2] >= cubeB[0][2]) and (cubeA[0][2] <= cubeB[1][2])

overlap([[1, 0, 1], [1, 2, 1]], [[0, 0, 1], [2, 0, 1]])
```

```python
from tqdm import tqdm
def move_blocks_down(blocks):
    # Sort blocks based on their starting Z coordinate in descending order
    sorted_blocks = sorted(blocks, key=lambda x: x[0][2], reverse=False)
    new_blocks = []
    falls = 0
    for i in range(len(sorted_blocks)):
        current_block = sorted_blocks[i]
        block_start = current_block[0]
        block_end = current_block[1]
        
        if block_start[2] == 1 or block_end[2] == 1:
            new_blocks.append(current_block)
            continue
        fall_bool = False
        while not (block_start[2] == 1 or block_end[2] == 1):
            block_start[2] -= 1
            block_end[2] -= 1
            current_block = [block_start, block_end]
            if any([overlap(current_block, below_block) for below_block in sorted_blocks[0:i]]):
                block_start[2] += 1
                block_end[2] += 1
                break
            fall_bool= True
        if fall_bool:
            falls += 1
   
    return sorted_blocks, falls

bricks, falls = move_blocks_down(bricks)
bricks, falls
```

```python
from collections import Counter
def are_3d_block_lists_equal(list1, list2):
    return Counter(tuple(map(tuple, sublist)) for sublist in list1) == Counter(tuple(map(tuple, sublist)) for sublist in list2)
```

```python
import copy
ans = 0
falls_ans = []
for i in tqdm(range(len(bricks))):
    dis_blocks = bricks[:i] + bricks[i + 1:]
    dis_blocks = copy.deepcopy(dis_blocks)
    old_dis_blocks = copy.deepcopy(dis_blocks)
    dis_blocks, falls = move_blocks_down(dis_blocks)
    falls_ans.append(falls)
    if are_3d_block_lists_equal(old_dis_blocks, dis_blocks):
        ans += 1
f"22-1: {ans}, 22-2: {sum(falls_ans)}"
```
