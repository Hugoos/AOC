```python
#data = "aaaaa-bbb-z-y-x-123[abxyz]"

data = """
123 -> x
456 -> y
x AND y -> d
x OR y -> e
x LSHIFT 2 -> f
y RSHIFT 2 -> g
NOT x -> h
NOT y -> i
"""
with open("input_7.txt", 'r') as f:
    data = f.read()
operations = []
for line in data.split("\n"):
    if line:
        operations.append(line.split(" -> "))

#print(operations)
```

```python
wires = {}

def getVal(x):
    if x.isalpha() and x in wires.keys():
        return wires[x]
    elif x.isalpha() and not x in wires.keys():
        return -1
    else:
        return int(x)
            
mask = 2 ** 16 - 1    

while len(operations) > 0:
    opp, target = operations.pop(0)
    if "AND" in opp:
        opp1,opp2 = opp.split(" AND ")
        opp1 = getVal(opp1)
        opp2 = getVal(opp2)
        if opp1 == -1 or opp2 == -1:
            operations.append([opp,target])
            continue
        wires[target] = (opp1 & opp2) & mask
        continue
    elif "OR" in opp:
        opp1,opp2 = opp.split(" OR ")
        opp1 = getVal(opp1)
        opp2 = getVal(opp2)
        if opp1 == -1 or opp2 == -1:
            operations.append([opp,target])
            continue
        wires[target] = (opp1 | opp2) & mask
        continue
    elif "LSHIFT" in opp:
        opp1,opp2 = opp.split(" LSHIFT ")
        opp1 = getVal(opp1)
        opp2 = getVal(opp2)
        if opp1 == -1 or opp2 == -1:
            operations.append([opp,target])
            continue
        wires[target] = (opp1 << opp2) & mask
        continue
    elif "RSHIFT" in opp:
        opp1,opp2 = opp.split(" RSHIFT ")
        opp1 = getVal(opp1)
        opp2 = getVal(opp2)
        if opp1 == -1 or opp2 == -1:
            operations.append([opp,target])
            continue
        wires[target] = (opp1 >> opp2) & mask
        continue
    elif "NOT" in opp:
        opp1 = opp.split("NOT ")[1]
        opp1 = getVal(opp1)
        if opp1 == -1:
            operations.append([opp,target])
            continue
        wires[target] = (~opp1) & mask
        continue
    else:
        opp1 = getVal(opp)
        if opp1 == -1:
            operations.append([opp,target])
            continue
        wires[target] = opp1
        continue
```
