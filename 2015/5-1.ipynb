```python
data = "dvszwmarrgswjxmb"
with open("input_5.txt", 'r') as f:
    data = f.read()
print(data)
data = [x for x in data.split("\n") if not x == ""]
print(data)
```

```python
num_nice = 0
num_naughty = 0
for word in data:
    num_vowels = len([x for x in word if x in list("aeiou")])
    num_equal = sum([x == y for x,y in zip(word[:-1], word[1:])])
    num_list = sum([x in word for x in ["ab", "cd", "pq", "xy"]])
    if num_vowels >= 3 and num_equal >= 1 and num_list == 0:
        num_nice += 1
    else:
        num_naughty += 1
print(f"""nice: {num_nice}\nnaughty: {num_naughty}""")
```
