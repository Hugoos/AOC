```python
import numpy as np
data = """
../.# => ##./#../...
.#./..#/### => #..#/..../..../#..#
"""
with open("input_22.txt", 'r') as f:
    data = f.read()

data = [x for x in data.split("\n") if not x == ""]

#data= data.replace("\n", "")
convertDict = {".": 0, "#": 1}
#grid = np.array([[0,0,1],[1,0,0],[0,0,0]])
data
grid = []
for row in data:
    r = []
    for char in row:
        r.append(convertDict[char])
    grid.append(r)
grid = np.array(grid)
grid 
```

```python
import math
move = {"N": (0,-1), "S": (0,1), "W": (-1,0), "E": (1,0)}
m_left = {"N": "W", "W": "S", "S": "E", "E": "N"}
m_right = {"N": "E", "W": "N", "S": "W", "E": "S"}
DYDX = [(1,0),(0,1), (-1,0), (0,-1)]
new_pos_dict = {(0,1): "S", (0,-1): "N", (-1,0): "E", (1,0): "W"}
state = {}
```

```python

for i,y in enumerate(range(math.ceil(-len(grid) /2), math.ceil(len(grid) /2))):
    for j,x in enumerate(range(math.ceil(-len(grid[0]) /2), math.ceil(len(grid[0]) /2))):
        state[(x,y)] = grid[i,j]
state
```

```python
def burst(pos,facing, bursts):
    numInfected = 0
    for x in range(bursts):
        current_node = state[pos]
        if current_node:
            facing = m_right[facing]
            state[pos] = 0
        elif not current_node:
            facing = m_left[facing]
            state[pos] = 1
            numInfected += 1
        else:
            print("error")
        pos = tuple(np.add(list(pos),list(move[facing])))
        if not pos in state:
            state[pos] = 0
    return numInfected    
```

```python
pos = (0,0)
facing = "N"
burst(pos, facing, 10000)
```
