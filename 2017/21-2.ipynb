```python
def fractal(frac_arr):
    #new_arr = []
    if frac_arr.shape[0] % 2 == 0:
        #split in 2x2 arrays
        frac_arr = resample(frac_arr, 2)
    elif frac_arr.shape[0] % 3 == 0:
        #split in 3x3 arrays
        frac_arr = resample(frac_arr, 3)
    return frac_arr
```

```python
def npNicePrint(arr):
    convertDict = {0: ".", 1: "#"}
    s = ""
    for r in arr:
        for c in r:
            s += convertDict[c]
        s += "\n"
    print(s)
def resample(arr, N):
    A = []
    for v in np.vsplit(arr, arr.shape[0] // N):
        A.extend([*np.hsplit(v, arr.shape[0] // N)])
    return np.array(A)
```

```python
def getAllConfigurations(arr):
    configurationList = [arr]
    configurationList.append(np.fliplr(arr))
    for i in range(3):
        arr = np.rot90(arr)
        configurationList.append(arr)
        configurationList.append(np.fliplr(arr))
    return configurationList

def checkLineUp(arr1, arr2):
    con1 = getAllConfigurations(arr1)
    #con2 = getAllConfigurations(arr2)
    for grid1 in con1:
        #print(grid1,arr2)
        if np.array_equal(grid1,arr2):
            return True
    return False

def transformTile(arr1):
    for tile in formulas:
        if checkLineUp(arr1,tile[0]):
            return tile[1]
```
