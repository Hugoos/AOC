```python
import numpy as np
data = """
0/2
2/2
2/3
3/4
3/5
0/1
10/1
9/10
"""
with open("input_24.txt", 'r') as f:
    data = f.read()

data = [x for x in data.split("\n") if not x == ""]

#data= data.replace("\n", "")

data
```

```python
portsList = []
for port in data:
    port = list(map(int,port.split("/")))
    portsList.append(port)
```

```python
portsList
```

```python
import copy

def connect(portsList, nextPort = 0, portsChain=[]):
    hits = 0
    #print(portsChain)
    for port in portsList:
        if nextPort in port:
            hits += 1
            port_c = copy.deepcopy(port)
            portsList_c = copy.deepcopy(portsList)
            portsList_c.remove(port_c)
            portsChain_c = copy.deepcopy(portsChain)
            portsChain_c.append(copy.deepcopy(port_c))
            #print(port,portsChain)
            configurations.append(portsChain_c)
            #print(port)
            port_c.remove(nextPort)   
            #print(portsList)
            #print(port)
            connect(portsList_c, port_c[0], portsChain_c)
        
```

```python
configurations = []
connect(portsList)
```

```python
configurations
```

```python
sumList = []
longest = []
for x in configurations:
    longest.append(len([x for y in x]))

longest = max(longest)
longest
longestList = []
longestList = [x for x in configurations if len(x) == longest]
longestList 
```

```python
for x in longestList:
    sumList.append(sum([sum(y) for y in x]))
sumList.sort(reverse=True)
print(sumList[0])
```
