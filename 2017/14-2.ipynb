```python
from functools import reduce
from operator import xor
def knot_hash(data_str):
    linked_list = {}
    l = [x for x in range(256)]
    data = []
    for char in data_str:
        data.append(ord(char))
    data.extend([17, 31, 73, 47, 23])  
    for i, digit in enumerate(l):
        linked_list[i] = {"data": int(digit), "next_i": (i+1)%(len(l))}

    current_position = 0
    skip_size = 0
    linked_list
    for x in range(64):
        for length in data:
            reversedList = list(reversed([linked_list[x]['data'] for x in [(current_position + y) % len(l) for y in range(length)]]))
            for i, x in enumerate([(current_position + y) % len(l) for y in range(length)]):
                linked_list[x]['data'] = reversedList[i]
            current_position = (current_position + length + skip_size) % len(l)
            skip_size += 1

    spare_hash_list = []
    solution = ""
    for i in range(256):
        spare_hash_list.append(linked_list[i]['data'])
        if i % 16 == 15:
            spare_hash_list
            solution += format(reduce(xor, map(int, spare_hash_list)), '02x')
            spare_hash_list = []
    return solution 
```

```python
import numpy as np
arr = np.zeros((128,128))
for x in range(128):
    knot = knot_hash(f"{puzzle_in}-{x}")
    stra = ""
    for char in knot:
        stra += '{:04b}'.format(int(char, 16))
    arr[x,:] = np.fromstring(stra, np.int8) - 48
```

```python
print(sum(sum(arr)))
```

```python
import networkx as nx
neighbours = [(-1,0),(1,0),(0,1),(0,-1)]
G = nx.Graph()
for y, row in enumerate(arr):
    #print(row)
    for x, elem in enumerate(row):
        if elem == 0:
            continue
        loc = (x,y)
        #print(loc)
        G.add_node(loc)
        #print(f"{loc}\n---------\n")
        for n in neighbours:
            loc2 = tuple(np.add(loc,n))
            #print(loc2)
            if not (loc2[0] < 0 or loc2[1] < 0 or loc2[0] >= len(row) or loc2[1] >= len(row)):
                if arr[loc2[1],loc2[0]] == 1:
                    #print(loc,loc2)
                    G.add_edge(loc,loc2)
```

```python
conn = nx.number_connected_components(G)
alone = len([node for node in G.nodes if G.degree(node) == 0])
print(conn, alone, conn+alone)
```

```python
sub_graphs = (G.subgraph(c) for c in nx.connected_components(G))
```

```python
summetje = 0
for graph in sub_graphs:
    summetje += len(graph.nodes)
    print(len(graph.nodes))
print(summetje)
```

```python
def npNicePrint(arr):
    convertDict = {0: ".", 1: "#"}
    s = ""
    for r in arr:
        for c in r:
            s += convertDict[c]
        s += "\n"
    print(s)
    return s
```
