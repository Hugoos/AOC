```python
with open("input_12.txt", 'r') as f:
    data = f.read()
#data = """
#0 <-> 2
#1 <-> 1
#2 <-> 0, 3, 4
#3 <-> 2, 4
#4 <-> 2, 3, 6
#5 <-> 6
#6 <-> 4, 5
#"""
#data= data.replace("\n", "")
data = [x for x in data.split("\n") if not x == ""]
data
```

```python
import networkx as nx

G = nx.Graph()
for line in data:
    groups = line.split(" <-> ")
    origin = int(groups[0])
    connected = [int(x) for x in groups[1].split(",")]
    G.add_node(origin)
    for neighbour in connected:
        G.add_edge(origin,neighbour)
```

```python
nx.draw_kamada_kawai(G)
```

```python
len(nx.node_connected_component(G, 0))
```
