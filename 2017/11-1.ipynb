```python
with open("input_11.txt", 'r') as f:
    data = f.read()
#data = "ne,ne,sw,sw"
data= data.replace("\n", "")
data = data.split(",")
data
```

```python
hexaDistance = {"se" : [1,-1,0],"s": [0,-1,1], "sw": [-1,0,1],
                "nw": [-1,1,0], "ne": [1,0,-1], "n": [0,1,-1]}
def cube_distance(a, b):
    return (abs(a[0] - b[0]) + abs(a[1] - b[1]) + abs(a[2] - b[2])) / 2
cube_distance((0,0,0), (1,-1,0))
```

```python
import numpy as np
position = (0,0,0)
for move in data:
    position = tuple(np.add(list(position), hexaDistance[move]))
cube_distance((0,0,0), position)
```
