```python
linked_list = {}
data = str(91212129)

with open("input_1.txt", 'r') as f:
    data = f.read()
data = data.split("\n")[0]
```

```python
for i, digit in enumerate(data):
    linked_list[i] = {"data": int(digit), "next_i": (i+1)%(len(data))}
```

```python
linked_list
```

```python
sum([linked_list[key]['data'] for key in linked_list.keys() if linked_list[key]['data'] == linked_list[linked_list[key]['next_i']]['data']])
    
```
