```python
data = """
     |          
     |  +--+    
     A  |  C    
 F---|----E|--+ 
     |  |  |  D 
     +B-+  +--+ 
"""
with open("input_19.txt", 'r') as f:
    data = f.read()

data = [list(x) for x in data.split("\n") if not x == ""]

#data= data.replace("\n", "")
```

```python
import copy
move = {"N": (0,1), "S": (0,-1), "W": (-1,0), "E": (1,0)}
DYDX = [(1,0),(0,1), (-1,0), (0,-1)]
new_pos_dict = {(0,1): "S", (0,-1): "N", (-1,0): "E", (1,0): "W"}
def getNeighbours(pos, cameFrom = None):
    n_DYDX = copy.deepcopy(DYDX)
    if cameFrom:
        n_DYDX.remove(move[cameFrom])
    n_list = []
    for dydx in n_DYDX:
        new_pos = (pos[0]+dydx[0],pos[1]+dydx[1])
        if -1 < new_pos[0] < len(data[0]) and -1 < new_pos[1] < len(data):
            char = data[new_pos[1]][new_pos[0]]
        else:
            continue
        if not char == " ":
            n_list.append([char,new_pos_dict[dydx], new_pos])
    return n_list
        
```

```python
moving = "S"
word = ""
x_first = [i for i,x in enumerate(data[0]) if not x == " "][0]
curr_pos = (x_first,0)
step = 0
while True:
    skip_round = False
    #print(curr_pos)
    for neigh in getNeighbours(curr_pos, moving):
        char = neigh[0]
        direction = neigh[1]
        if direction == moving:
            curr_pos = neigh[2]
            if char.isalpha():
                word += char
            skip_round = True
            step += 1
            break
    if skip_round:
        continue
    for neigh in getNeighbours(curr_pos, moving):
        char = neigh[0]
        moving = neigh[1]
        curr_pos = neigh[2]
        
        if char.isalpha():
            word += char
        skip_round = True
        step += 1
        break
    if skip_round:
        continue
    print(word)
    step += 1
    print(step)
    break
        
```
