```python
data = """
pbga (66)
xhth (57)
ebii (61)
havc (66)
ktlj (57)
fwft (72) -> ktlj, cntj, xhth
qoyq (66)
padx (45) -> pbga, havc, qoyq
tknk (41) -> ugml, padx, fwft
jptl (61)
ugml (68) -> gyxo, ebii, jptl
gyxo (61)
cntj (57)
"""
with open("input_7.txt", 'r') as f:
    data = f.read()
data = [x for x in data.split("\n") if not x == ""]
```

```python
import networkx as nx
G = nx.DiGraph()
for line in data:
    line = line.replace("-> ", "")
    line = line.replace(",", "")
    line = line.replace("(", "")
    line = line.replace(")", "")
    line = line.split(" ")
    prog = line[0]
    weight = line[1]
    children = line[2:]
    G.add_node(prog, weight=weight)
    for child in children:
        G.add_edge(prog, child)
```

```python
nx.draw_kamada_kawai(G, with_labels=True, font_weight='bold')
print([x for x in G.nodes() if G.in_degree(x) == 0][0])
```
