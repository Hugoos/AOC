Afstand van coördinaat tot coördinaat;
d(p,q) = v- (p1-q1)^2 + (p2-q2)^2 + (p3-q3)^2

```python
from scipy.spatial.distance import cdist

D = cdist(data, data, metric="euclidean")
# upper-triangular indices (i < j)
mask = np.triu(np.ones_like(D, dtype=bool), k=1)
D[mask] = np.inf
D[D == 0] = np.inf

k = 1000   # how many smallest

flat = D.ravel()
idx = np.argpartition(flat, k)[:k]

# sort the selected ones
order = np.argsort(flat[idx])

idx_sorted = idx[order]
coords = np.column_stack(np.unravel_index(idx_sorted, D.shape))
values = flat[idx_sorted]

#coords
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_nodes_from(data)

for coord in coords:
    val1 = data[coord[0]]
    val2 = data[coord[1]]
    if not nx.has_path(G, val1, val2):
        G.add_edge(val1, val2)

#pos = nx.kamada_kawai_layout(G)    # computes positions
#nx.draw(G, pos, with_labels=True, node_size=100)

#plt.show()
```

```python
components = list(nx.connected_components(G))

comps_sorted = sorted(components, key=len, reverse=True)

x = 3   # how many largest you want
largest = comps_sorted[:x]
largest

ans = 1

for x in largest:
    ans *= len(x)
ans
```
