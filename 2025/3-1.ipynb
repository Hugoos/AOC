```python
def find_max(row):
    i_1 = np.argmax(row[:-1]) 
    i_2 = np.argmax(row[i_1+1:]) 
    return int(str(row[i_1]) + str(row[i_1+1:][i_2]))

sum([find_max(row) for row in data])
```
