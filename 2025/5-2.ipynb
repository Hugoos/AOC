```python
import helpers
import numpy as np

data = """3-5
10-14
16-20
12-18

1
5
8
11
17
32"""

with open("input_5.txt", 'r') as f:
    data = f.read()
    
# Split on the double newline
first, second = data.strip().split("\n\n")

# Parse ranges
ranges = []
for line in first.splitlines():
    a, b = map(int, line.split("-"))
    ranges.append([a, b])

# Parse trailing numbers
numbers = [int(x) for x in second.splitlines()]
```

```python
#inefficient with no sort
combined_ranges = [ranges[0]]

while True:
    combined_ranges = [ranges[0]]
    for r in ranges[1:]:
        combined = False
        for r2 in combined_ranges:
            #case 1 range is completely inside
            if (r2[0] <= r[0] <= r2[1]) and (r2[0] <= r[1] <= r2[1]):
                combined = True
                break #range can be discarded
            #case 2 range start is lower than original start, range end is inside
            if r[0] < r2[0] and r2[0] <= r[1] <= r2[1]: 
                combined = True
                r2[0] = r[0]
                break
            #case 3 range start is inside, range end is higher than original start
            if r2[0] <= r[0] <= r2[1] and r[1] > r2[1]:  
                combined = True
                r2[1] = r[1]
                break
            #case 4 range start is lower than original and range end is higher than original
            if r[0] < r2[0] and r[1] > r2[1]:
                combined = True
                r2[0] = r[0]
                r2[1] = r[1]
                break
        if not combined:
            combined_ranges.append(r) #new range added
    if ranges == combined_ranges:
        break
    ranges = combined_ranges
    
ans = 0
for r in combined_ranges:
    ans += r[1] - r[0] + 1
ans
```

```python
#efficient
ranges.sort(key=lambda x: x[0])
combined_ranges = [ranges[0]]
for new in ranges[1:]:
    last = combined_ranges[-1]
    if new[0] <= last[1]:
        last[1] = max(last[1], new[1])
    else:
        combined_ranges.append(new)
ans = 0
for r in combined_ranges:
    ans += r[1] - r[0] + 1
ans
```

```python
#with reduce, one line
from functools import reduce
ranges.sort(key=lambda x: x[0])

def merge_intervals(merged, current):
    if not merged or merged[-1][1] < current[0] - 1:
        merged.append(current)
    else:
        merged[-1][1] = max(merged[-1][1], current[1])
    return merged

sum([r[1] - r[0] + 1 for r in reduce(merge_intervals, ranges, [])])
```
