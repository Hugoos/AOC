```python
import helpers
import numpy as np

data = """123 328  51 64 
 45 64  387 23 
  6 98  215 314
*   +   *   +  """

with open("input_6.txt", 'r') as f:
    data = f.read()
    
data = data.split("\n")
```

```python
problem = [[] for x in range(len(data))]
ans = 0
for i in range(len(data[0])):
    for j in range(len(data)):
        problem[j].append(data[j][i])
    if len(set([problem[x][-1] for x in range(len(data))])) == 1:
        nums = [str(int("".join([problem[j][i] for j in range(len(problem)-1)]).strip())) for i in range(len(problem[0])-1)]
        ans += eval(f"{problem[-1][0]}".join(nums)) 
        problem = [[] for x in range(len(data))] 

nums = [str(int("".join([problem[j][i] for j in range(len(problem)-1)]).strip())) for i in range(len(problem[0]))]
ans += eval(f"{problem[-1][0]}".join(nums)) 
ans       
```
