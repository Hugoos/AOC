```python
import helpers
import numpy as np

#helpers.npNicePrint(arr)

data = """987654321111111
811111111111119
234234234234278
818181911112111
"""
with open("input_3.txt", 'r') as f:
    data = f.read()
    
#changeDict = {"forward": np.array([1,0]), "down": np.array([0,1]), "up": np.array([0,-1])}
#data = [changeDict[x.split(" ")[0]] * int(x.split(" ")[1]) for x in data.split("\n") if x]
#data = [x.split(' ') for x in data.split('\n') if x]
#np.prod(sum(data))    
data = [x for x in data.split('\n') if not x == ""]
data = [[int(y) for y in x] for x in data ]
#data
```

```python
#iterative solution
def find_max(row, num_batteries):
    ans = ""
    i = 0
    for x in range(num_batteries):
        back_i = -num_batteries + 1 + x #we can't take any that makes the row a lower length than the amount of digits left
        if not back_i == 0:
            i = np.argmax(row[:back_i]) 
            highest_num = row[i]
            row = row[i+1:]
        else: #if this is 0 then row[:0] would give the incorrect row
            i = np.argmax(row) 
            highest_num = row[i]
        ans += str(highest_num)    
    return int(ans)

print(f"part 1: {sum([find_max(row, 2) for row in data])}")
print(f"part 2: {sum([find_max(row, 12) for row in data])}")
```

```python
#recursive solution
def find_max_r(row, num_batteries):
    if 0 < len(row) < num_batteries or num_batteries < 1: #invalid inputs
        return np.nan
    if num_batteries > 1:
        i = np.argmax(row[:-num_batteries+1]) 
        ans = str(row[i]) + str(find_max_r(row[i+1:], num_batteries-1))
    elif num_batteries == 1: #base case
        ans = row[np.argmax(row)]
    return int(ans)

print(f"part 1: {sum([find_max_r(row, 2) for row in data])}")
print(f"part 2: {sum([find_max_r(row, 12) for row in data])}")
print(f"bonus (50): {sum([find_max_r(row, 50) for row in data])}")
print(f"bonus (99): {sum([find_max_r(row, 99) for row in data])}")
print(f"bonus (100): {sum([find_max_r(row, 100) for row in data])}")
print(f"bonus (101): {sum([find_max_r(row, 101) for row in data])}")
```
