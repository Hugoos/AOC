```python
change_dict = {
    "R": "+",
    "L": "-"
}
#eval is unsafe :D
instructions = [eval(change_dict[x[0]] + x[1:]) for x in instructions]
#instructions
```

```python
i = 50
ans = 0
for instruction in instructions:
    while not instruction == 0: 
        if instruction < 0:
            instruction += 1
            i = (i - 1) % 100
        elif instruction > 0:
            instruction -= 1
            i = (i + 1) % 100
        
        if i == 0:
            ans += 1
ans
```
