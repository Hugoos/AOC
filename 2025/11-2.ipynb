```python
import helpers
import numpy as np
from collections import defaultdict
import re

data = """svr: aaa bbb
aaa: fft
fft: ccc
bbb: tty
tty: ccc
ccc: ddd eee
ddd: hub
hub: fff
eee: dac
dac: fff
fff: ggg hhh
ggg: out
hhh: out"""

#with open("input_11.txt", 'r') as f:
#    data = f.read().rstrip()

connections = {}

for row in data.split("\n"):
    stuff = row.split(": ")
    node_list = []
    for connected_node in stuff[1].split(" "):
        node_list.append(connected_node)
    connections[stuff[0]] = node_list
```

```python
import networkx as nx

G = nx.DiGraph()

# Add edges
for src, targets in connections.items():
    for t in targets:
        G.add_edge(src, t)

# Check DAG
is_dag = nx.is_directed_acyclic_graph(G)
print("Is DAG:", is_dag)

# If not a DAG, show cycles
if not is_dag:
    print("Cycles:", list(nx.simple_cycles(G)))
```
