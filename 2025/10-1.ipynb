```python
import helpers
import numpy as np
from collections import defaultdict
import re

data = """[.##.] (3) (1,3) (2) (2,3) (0,2) (0,1) {3,5,4,7}
[...#.] (0,2,3,4) (2,3) (0,4) (0,1,2) (1,2,3,4) {7,5,12,7,2}
[.###.#] (0,1,2,3,4) (0,3,4) (0,1,2,4,5) (1,2) {10,11,11,5,10,5}"""

with open("input_10.txt", 'r') as f:
    data = f.read().rstrip()

def parse_line(line):
    # []
    bracket_str = re.search(r"\[(.*?)\]", line).group(1)
    first = tuple(1 if c == '#' else 0 for c in bracket_str)

    # ()
    paren_parts = re.findall(r"\((.*?)\)", line)
    second = [tuple(map(int, p.split(','))) for p in paren_parts]

    # {}
    curly_str = re.search(r"\{(.*?)\}", line).group(1)
    third = list(map(int, curly_str.split(',')))

    return first, second, third

data = [
    parse_line(line)
    for line in data.splitlines()
    if line.strip()
]
```

```python
from collections import deque

def toggle_machine(machine, move):
    machine = list(machine)
    for m in move:
        machine[m] = int(not machine[m])
    return tuple(machine)

def bfs(state, moves, target):
    visited = set()
    queue = deque([(state, 0)])
    visited.add(state)

    while queue:
        state, depth = queue.popleft()
        if state == target:
            return depth

        for move in moves:
            new_state = toggle_machine(state, move)
            if new_state not in visited:
                visited.add(new_state)
                queue.append((new_state, depth+1))

ans = 0
for target, moves, joltage in data:
    start_state = (0,) * len(target)
    ans += bfs(start_state, moves, target)
ans
```
