```python
import helpers
import numpy as np

data = """3-5
10-14
16-20
12-18

1
5
8
11
17
32"""

with open("input_5.txt", 'r') as f:
    data = f.read()
    
# Split on the double newline
first, second = data.strip().split("\n\n")

# Parse ranges
ranges = []
for line in first.splitlines():
    a, b = map(int, line.split("-"))
    ranges.append((a, b))

# Parse trailing numbers
numbers = [int(x) for x in second.splitlines()]
```

```python
ans = 0
for num in numbers:
    for r in ranges:
        if r[0] <= num <= r[1]:
            ans += 1
            break
ans        
```
