```python
import helpers
import numpy as np
from collections import defaultdict

data = """.......S.......
...............
.......^.......
...............
......^.^......
...............
.....^.^.^.....
...............
....^.^...^....
...............
...^.^...^.^...
...............
..^...^.....^..
...............
.^.^.^.^.^...^.
..............."""

with open("input_7.txt", 'r') as f:
    data = f.read()
    
# Split on the double newline
grid = defaultdict(lambda: None, {(x,y): val for y, r in enumerate(data.split('\n')) for x, val in enumerate(r)})
#helpers.dictNicePrint(grid)
```

```python
#(c,r) down is +, up is -

S_pos = next((pos for pos, val in grid.items() if val == "S"), None)

accessed = set()

ans = 0
def descend(pos):
    global ans
    global accessed
    accessed.add(pos)
    next_pos = (pos[0], pos[1]+1)
    if not next_pos in grid.keys() or next_pos in accessed:
        return
    if grid[next_pos] == ".":
        return descend(next_pos)
    elif grid[next_pos] == "^":
        ans += 1
        descend((next_pos[0]-1, next_pos[1]))
        descend((next_pos[0]+1, next_pos[1]))

descend(S_pos)

ans
```
