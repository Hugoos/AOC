```python
numbers = data[0].split(",")
numbers
```

```python
class grid:
    def __init__(self, vals):
        vals = [list(filter(None, x.split(" "))) for x in vals.split("\n") ]
        self.bingo_grid = np.array(vals).astype(int)
        self.cleared_grid = np.zeros_like(self.bingo_grid)
        
    def drawNumber(self, number):
        self.cleared_grid[np.where(self.bingo_grid == number)] = 1
        
    def checkWin(self):
        for x in self.cleared_grid:
            if np.all(x):
                return True
        for x in self.cleared_grid.T:
            if np.all(x):
                return True
        return False
    
    def calculateScore(self):
        return sum(self.bingo_grid[np.where(self.cleared_grid == 0)])
        
```

```python
grid_list = []
for g in data[1:]:
    grid_list.append(grid(g))
```

```python
final_number = -1
winning_board = None
finished = False
for number in numbers:
    for g in grid_list:
        g.drawNumber(int(number))
        if g.checkWin():
            final_number = int(number)
            winning_board = g
            print(f"{winning_board} won with {final_number}")
            finished = True
            break
    if finished:
        break
        
    
winning_board.calculateScore() * final_number
```
