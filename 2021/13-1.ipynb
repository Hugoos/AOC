```python
def nprintCoords(coord):
    x_s = 0
    y_s = 0
    for x,y in coord:
        if x > x_s:
            x_s = x
        if y > y_s:
            y_s = y
    arr = np.zeros((y_s+1,x_s+1))
    for x,y in coord:
        arr[y,x] = 1
    helpers.npNicePrint(arr)
```

```python
for instruction in instructions:
    fold_type, fold = instruction.split("=")
    fold = int(fold)
    newCoords = set()
    print(fold, fold_type)
    for x,y in coords:
        if y < fold and fold_type == "y":
            newCoords.add((x,y))
        if y > fold and fold_type == "y":
            newCoords.add((x, 2*fold - y))
        if x < fold and fold_type == "x":
            newCoords.add((x,y))
        if x > fold and fold_type == "x":
            newCoords.add((2*fold-x,y))
    print(len(newCoords))
    nprintCoords(newCoords)
    coords = newCoords
    
```
