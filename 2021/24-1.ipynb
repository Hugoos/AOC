```python
class ALU():
    def __init__(self):
        self.v = {"w": 0, "x": 0, "y": 0, "z": 0 }
        self.inputQ = []
    
    def initialize(self):
        self.v = {"w": 0, "x": 0, "y": 0, "z": 0 }
        
    def getVal(self, val):
        if val in self.v.keys():
            return int(self.v[val])
        else:
            return int(val)
        
    def execute_line(self, line):
        if line[0] == "inp":
            if len(self.inputQ) > 0:
                self.v[line[1]] = self.inputQ.pop(0)
            else:
                self.v[line[1]] = int(input())
        elif line[0] == "add":  
            self.v[line[1]] += self.getVal(line[2])
        elif line[0] == "mul":
            self.v[line[1]] *= self.getVal(line[2])
        elif line[0] == "div":
            self.v[line[1]] = math.floor(self.v[line[1]] / self.getVal(line[2]))
        elif line[0] == "mod":
            self.v[line[1]] %= self.getVal(line[2])
        elif line[0] == "eql":
            self.v[line[1]] = self.v[line[1]] == self.getVal(line[2])
        else:
            print("error")
    
    def execute(self, data):
        self.initialize()
        data = [x.split(" ") for x in data.split("\n")]
        for l in data:
            self.execute_line(l)
            
    def setInput(self, inp):
        self.inputQ = inp
```

```python
import sys

def run(z, d, c1, c2):
    cond = (z % 26 + c1) != d
    if c1 < 0:
        z //= 26
    if cond:
        z = 26 * z + d + c2

    return z

zs = {0: ()}
for idx, line in enumerate(sys.stdin):
    c1, c2 = map(int, line.split())
    CAP = 26**4
    zs = {
        run(z, d, c1, c2): v + (d,)
        for z, v in zs.items() for d in range(1, 10) # reverse range for part 2
        if z <= CAP
    }

print("".join(str(c) for c in zs[0]))
```
