```python
from scipy.spatial import Delaunay
def manhattan(a, b):
    return sum(abs(val1-val2) for val1, val2 in zip(a,b))
```

```python
def sameArea(b1,b2):
    b1_d = sorted([manhattan(x,y) for x in b1 for y in b1 if not x == y])
    b2_d = sorted([manhattan(x,y) for x in b2 for y in b2 if not x == y])
    return b1_d == b2_d

def match(b1,b2):
    for b2_p in it.permutations(scanners[1], 3):
        lst = [manhattan(x,y) for x,y in zip(b1,b2_p)]
        if all(ele == lst[0] for ele in lst):
            return b1, b2_p
```

```python
def Rx(theta):
    return np.matrix([[ 1, 0           , 0           ],
                   [ 0, math.cos(theta),-math.sin(theta)],
                   [ 0, math.sin(theta), math.cos(theta)]])
  
def Ry(theta):
    return np.matrix([[ math.cos(theta), 0, math.sin(theta)],
                   [ 0           , 1, 0           ],
                   [-math.sin(theta), 0, math.cos(theta)]])
  
def Rz(theta):
    return np.matrix([[ math.cos(theta), -math.sin(theta), 0 ],
                   [ math.sin(theta), math.cos(theta) , 0 ],
                   [ 0           , 0            , 1 ]])

v = [264, 108, 156]
axis = [0, 0, 0]
theta = math.radians(270) 

def beacons_permutation(b):
    permutations = []
    for theta in [math.radians(0),math.radians(90),math.radians(180),math.radians(270)]:
        rot_x = Rx(theta)
        for theta2 in [math.radians(0), math.radians(90),math.radians(180),math.radians(270)]:
            rot_y = Ry(theta2)
            new_perm = []
            for v in b:
                nn = np.dot(rot_x, v).tolist()[0]
                new_perm.append([int(round(x)) for x in np.dot(rot_y, nn).tolist()[0]])
            permutations.append(new_perm)
    
    for theta in [math.radians(90),math.radians(270)]:
        rot_z = Rz(theta)
        for theta2 in [math.radians(0), math.radians(90),math.radians(180),math.radians(270)]:
            rot_y = Ry(theta2)
            new_perm = []
            for v in b:
                nn = np.dot(rot_z, v).tolist()[0]
                new_perm.append([int(round(x)) for x in np.dot(rot_y, nn).tolist()[0]])
            permutations.append(new_perm)    
    return permutations



test_s = set()
for x in beacons_permutation([[264, 108, 156]]):
    for y in x:
        test_s.add(tuple(y))
```
