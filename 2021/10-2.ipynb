```python
brackets = {'[': 0, '(': 0, '{': 0, '<': 0, 
            ']': 0, ')': 0, '}': 0, '>': 0}
score = {')': 3, ']': 57, '}': 1197, '>': 25137}

opening = ['[', '(', '{', '<']
closing = [']', ')', '}', '>']
matching = {'[': ']', '(': ')', '{': '}', '<': '>'}
lastOpened = []

def checkValid(brackets):
    if brackets['['] >= brackets[']'] and brackets['('] >= brackets[')'] and brackets['{'] >= brackets['}'] and brackets['<'] >= brackets['>']: 
        return True
    else:
        return False
data2 = []
points = 0    
for r in data:
    lastOpened = []
    add = True
    for char in r:
        if char in opening:
            lastOpened.append(char)
        elif char in closing:
            matchingChar = lastOpened.pop()
            if not char == matching[matchingChar]:
                points += score[char]
                add = False
                break
    if add:
        data2.append(r)
data2            
        
```

```python
score2 = {')': 1, ']': 2, '}': 3, '>': 4}
scoresList = []
for r in data2:
    lastOpened = []
    completionString = ""
    for char in r:
        if char in opening:
            lastOpened.append(char)
        elif char in closing:
            matchingChar = lastOpened.pop()
    while lastOpened:
        char = lastOpened.pop()
        completionString += matching[char]
    points = 0
    for char in completionString:
        points = points * 5
        points += score2[char]
    print(completionString, points)
    scoresList.append(points)
scoresList.sort()
print(scoresList)
print(scoresList[int((len(scoresList)-1)/2)])
```
