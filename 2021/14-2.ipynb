```python
import helpers
import numpy as np
import re
import math
from tqdm import tqdm
import copy
import networkx as nx
from collections import Counter

data = """NNCB

CH -> B
HH -> N
CB -> H
NH -> C
HB -> C
HC -> B
HN -> C
NN -> C
BH -> H
NC -> B
NB -> B
BN -> B
BB -> N
BC -> B
CC -> N
CN -> C"""

with open("input_14.txt", 'r') as f:
    data = f.read()
#changeDict = {"forward": np.array([1,0]), "down": np.array([0,1]), "up": np.array([0,-1])}
#data = re.findall("(\d+),(\d+) -> (\d+),(\d+)", data)
#data = [x.rstrip() for x in data.split("\n\n") if x]


data = data.lstrip()
data = data.rstrip()
polymer, instructions = data.split("\n\n")
instructions = [x.split(" -> ")  for x in instructions.split("\n")]

#data = [x.split("-") for x in data.split("\n")]
polymer, instructions
rules = {}
counts = {}
doubles = {}
for x,y in instructions:
    rules[x] = [x[0] + y,y + x[1]]
    counts[x] = 0
    doubles[y] = 0
emptyCounts = copy.deepcopy(counts)
emptyDoubles = copy.deepcopy(doubles)
for char1, char2 in zip(polymer,polymer[1:]):
    counts[char1+char2] += 1
polymer, rules, counts, doubles
```
