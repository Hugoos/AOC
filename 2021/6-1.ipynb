```python
def tick(data):
    num_new = len([x for x in data if x == 0])
    data = [x-1 for x in data]
    data = [x if not x == -1 else 6 for x in data]
    for x in range(num_new):
        data.append(8)
    return data
```

```python
for x in range(80):
    data = tick(data)
print(len(data))
```
