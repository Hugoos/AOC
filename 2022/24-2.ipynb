```python
import helpers
import numpy as np
import math
import re
from copy import deepcopy

data = """#.######
#>>.<^<#
#.<..<<#
#>v.><>#
#<^v^^>#
######.#"""


#with open("input_24.txt", 'r') as f:
#    data = f.read().rstrip()

state = []
for y, row in enumerate(data.split('\n')[1:-1]):
    for x, val in enumerate(row[1:-1]):
        if val == ">":
            state.append(((x,y), (1,0)))
        elif val == "<":
            state.append(((x,y), (-1,0)))
        elif val == "v":
            state.append(((x,y), (0,1)))
        elif val == "^":
            state.append(((x,y), (0,-1)))
            
height = len(data.split('\n')) - 2
width = len(data.split('\n')[0]) - 2
start = ((0, -1), 0)
end = (len(data.split('\n')[0])-3, len(data.split('\n'))-2)        
#locations = set([tuple(x) for x in data])
#locations
start, end
```

```python

def move_state(state):
    new_state = []
    for blizzard, movement in state:
        blizzard = np.array(blizzard) + np.array(movement)
        blizzard[0] = blizzard[0] % width
        blizzard[1] = blizzard[1] % height
        new_state.append((tuple(blizzard), movement))
    return new_state
```

```python
i = 0
states = {i: state}
new_state = None
while not states[0] == new_state:
    if i % 100 == 0:
        print(i)
    i += 1
    states[i] = move_state(states[i-1])
    new_state = states[i]
del states[i]
    
```

```python
way_5 = [np.array((0,0)), np.array((0,-1)),np.array((0,1)), np.array((1,0)),np.array((-1,0))]

def check_neighbours(coord, way, pos):
    #print(coord)
    neighs = [c for c in set([tuple(np.array(coord) + x) for x in way]) if not c in [x[0] for x in pos]]
    return [c for c in neighs if (0 <= c[0] < width and 0 <= c[1] < height) or c == end or c == start[0]]
```

```python
from collections import deque

reachable = set({})

def bfs(start):
    queue = deque([start])
    visited = set()

    while queue:
        current = queue.popleft()
        reachable.add(current)
        #print(current)
        if current[0] == end:
            
            return True  # Goal found
        next_i = (current[1]+1) % len(states)
        for neighbor in check_neighbours(current[0], way_5, states[next_i]):
            if neighbor not in visited:
                queue.append((neighbor, next_i))
                visited.add((neighbor, next_i))
    return False  # Goal not found            
```

```python
import heapq

def manhattan(pos1, pos2):
    #print(pos1, pos2)
    # Calculate the Manhattan distance between two positions
    x1, y1 = pos1
    x2, y2 = pos2
    return abs(x1 - x2) + abs(y1 - y2)

def astar(start, end):
    # Set of nodes already evaluated
    closed_set = set()

    # Set of currently discovered nodes that are not evaluated yet.
    # Initially, only the start node is known.
    open_set = {start}

    # For each node, which node it can most efficiently be reached from.
    # If a node can be reached from many nodes, came_from will eventually contain the
    # most efficient previous step.
    came_from = {}

    # For each node, the cost of getting from the start node to that node.
    g_score = {start: 0}

    # For each node, the total cost of getting from the start node to the goal
    # by passing by that node. That value is partly known, partly heuristic.
    f_score = {start: manhattan(start[0], end)}

    while open_set:
        current = min(open_set, key=lambda x: f_score[x])
        #print(current)

        if current[0] == end:
            path = []
            while current in came_from:
                path.append(current)
                current = came_from[current]
            path.append(start)
            return path[::-1]

        open_set.remove(current)
        closed_set.add(current)
        
        next_i = (current[1]+1) % len(states)
        for neighbor in check_neighbours(current[0], way_5, states[next_i]):
            if (neighbor, next_i) in closed_set:
                continue

            tentative_g_score = g_score[current] + 1

            if (neighbor, next_i) not in open_set:
                open_set.add((neighbor, next_i))
            elif tentative_g_score >= g_score[(neighbor, next_i)]:
                continue

            came_from[(neighbor, next_i)] = current
            g_score[(neighbor, next_i)] = tentative_g_score
            f_score[(neighbor, next_i)] = tentative_g_score + manhattan(neighbor, end)

    return []

def neighbors(grid, pos):
    # Return the neighbors of a given position in a grid
    x, y = pos
    neighbors = [(x+1, y), (x, y+1), (x-1, y), (x, y-1)]
    return [n for n in neighbors if n in grid]

# Example usage
#grid = {(0, 0), (0, 1), (1, 0), (1, 1)}
#start = (0, 0)
#end = (1, 1)
#path = astar(grid, start, end)
#print(path)  # [(0, 0), (1, 0), (1, 1)]
```
