```python
import helpers
import numpy as np

data = """A Y
B X
C Z
"""
with open("input_2.txt", 'r') as f:
    data = f.read()
    
#changeDict = {"forward": np.array([1,0]), "down": np.array([0,1]), "up": np.array([0,-1])}
#data = [changeDict[x.split(" ")[0]] * int(x.split(" ")[1]) for x in data.split("\n") if x]
data = [x.split(' ') for x in data.split('\n') if x]
#np.prod(sum(data))    
data
#A = Rock, B = Paper, C = Scissors
#X lose, Y = Draw, Z = Win
```

```python
beats = {'A': 'C', 'B': 'A', 'C': 'B'}
translate = {'X': beats, 'Y': {x: x for x in ['A', 'B', 'C']}, 'Z': {v: k for k, v in beats.items()}}
points_dict = {'A': 1, 'B': 2, 'C': 3}
beats = {'A': 'C', 'B': 'A', 'C': 'B'}
```

```python
points = 0
for game in data:
    move_enemy = game[0]
    move_me = translate[game[1]][move_enemy]
    points += points_dict[move_me]
    if move_enemy == move_me:
        points += 3
    elif move_enemy == beats[move_me]:
        points += 6
points
```
