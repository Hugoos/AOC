```python
from functools import partial

def move_block(cmb, movement):
    return [x + movement for x in cmb]

move_block_down = partial(move_block, movement=np.array((0, -1)))
```

```python
def check_overlap(cmb):
    if min(cmb,key=lambda item:item[0])[0] < 0 or max(cmb,key=lambda item:item[0])[0] >= 7 or min(cmb,key=lambda item:item[1])[1] < 0:
        return True
    if any([tuple(x) in settled for x in cmb]):
        return True
    return False
```
