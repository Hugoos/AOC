```python
def find_shortest_path(valley_map):
  # Initialize the starting position of the expedition
  start_row, start_col = None, None
  for i in range(len(valley_map)):
    if 'E' in valley_map[i]:
      start_row = i
      start_col = valley_map[i].index('E')
      valley_map[i] = valley_map[i].replace('E', '.')
      break
  print(valley_map)
  # Initialize the queue with the starting position
  queue = [(start_row, start_col, 0, valley_map)]
  visited = set()

  # BFS loop
  while queue:
    # Get the current position, the number of steps needed to reach it, and the current valley map
    row, col, steps, valley_map = queue.pop(0)
    print(valley_map)
    # Check if the current position is the destination
    if row == len(valley_map) - 1:
      return steps

    # Generate the next moves
    next_moves = [(row+1, col), (row-1, col), (row, col+1), (row, col-1)]
    
    # Iterate through the next moves
    for next_row, next_col in next_moves:
      # Check if the move is valid (inside the map and not a wall or a blizzard)
      if 0 <= next_row < len(valley_map) and 0 <= next_col < len(valley_map[0]) and valley_map[next_row][next_col] != '#' and (next_row, next_col) not in visited:
        # Update the valley map with the new blizzard positions
        new_valley_map = update_blizzards(valley_map)
        # Add the new position, the updated valley map, and the number of steps to the queue
        queue.append((next_row, next_col, steps+1, new_valley_map))
        visited.add((next_row, next_col))
  
  # If the destination was not reached, return -1
  return -1

def update_blizzards(valley_map):
  # Initialize a new valley map to store the updated positions
  new_valley_map = []
  print(valley_map)
  for row in valley_map:
    new_row = ''
    for col in row:
      new_row += col
    new_valley_map.append(new_row)

  # Iterate through the valley map and update the positions of the blizzards
  for i in range(len(valley_map)):
    for j in range(len(valley_map[0])):
      if valley_map[i][j] == '>':
        new_valley_map[i] = new_valley_map[i][:j] + '.' + new_valley_map[i][j+1:]
        new_valley_map[i] = new_valley_map[i][:j+1] + '>' + new_valley_map[i][j+2:]
      elif valley_map[i][j] == '<':
        new_valley_map[i] = new_valley_map[i][:j] + '.' + new_valley_map[i][j+1:]
        new_valley_map[i] = new_valley_map[i][:j-1] + '<' + new_valley_map[i][j:]
      elif valley_map[i][j] == '^':
        new_valley_map[i] = new_valley_map[i][:j] + '.' + new_valley_map[i][j+1:]
        if i == 0:
          new_valley_map[-1] = new_valley_map[-1][:j] + '^' + new_valley_map[-1][j+1:]
        else:
          new_valley_map[i-1] = new_valley_map[i-1][:j] + '^' + new_valley_map[i-1][j+1:]
      elif valley_map[i][j] == 'v':
        new_valley_map[i] = new_valley_map[i][:j] + '.' + new_valley_map[i][j+1:]
        if i == len(valley_map)-1:
          new_valley_map[0] = new_valley_map[0][:j] + 'v' + new_valley_map[0][j+1:]
        else:
          new_valley_map[i+1] = new_valley_map[i+1][:j] + 'v' + new_valley_map[i+1][j+1:]
    print(new_valley_map)
    return new_valley_map
        
# Example usage
data = """#E######
#>>.<^<#
#.<..<<#
#>v.><>#
#<^v^^>#
######.#"""
#valley_map = [  '#E#####',  '#.....#',  '#>....#',  '#.....#',  '#...v.#',  '#.....#',  '#####.#']
valley_map = data.split("\n")
shortest_path = find_shortest_path(valley_map)
print(shortest_path) # should print 4
```
