```python
class machine():
    def __init__(self, data):
        self.data = data
        self.ptr = 0
        self.cycle = 0
        self.executing_time = 0
        self.registers = {'X': 1}
        self.signal_strength = 0
    def run_cycle(self):
        if self.ptr >= len(self.data):
            return False
        self.cycle += 1
        if (self.cycle - 20) % 40 == 0:
            self.signal_strength += self.cycle * self.registers['X']
        #print(self.cycle, self.registers['X'], self.executing_time)
        if self.executing_time == 0:
            command = self.data[self.ptr]
            if command[0] == 'noop':
                self.executing_time = 1
            elif command[0] == 'addx':
                self.executing_time = 2
        self.executing_time -= 1
        if self.executing_time == 0:
            command = self.data[self.ptr]
            if command[0] == 'noop':
                self.ptr += 1
            elif command[0] == 'addx':
                self.ptr += 1
                self.registers['X'] += int(command[1])
        return True
```
