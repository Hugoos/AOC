```python
import helpers
import numpy as np
import math
import re
from copy import deepcopy

data = """>>><<><>><<<>><>>><<<>>><<<><<<>><>><<>>"""
with open("input_17.txt", 'r') as f:
    data = f.read().rstrip()

moves = {'>': np.array((1,0)), '<': np.array((-1,0))}
    
blocks = """####

.#.
###
.#.

..#
..#
###

#
#
#
#

##
##"""

blocks = [x.split('\n') for x in blocks.split('\n\n')]
blocks
blocks_coords = []

for block in blocks:
    block_coord = []
    start = np.array([0,0])
    for row in reversed(block):
        start[0] = 0
        for char in row:
            if char == "#":
                block_coord.append(deepcopy(start))
            start += np.array([1,0])
        start[1] += 1
    blocks_coords.append(block_coord)
blocks_coords    
data
```

```python
from functools import partial

def move_block(cmb, movement):
    return [x + movement for x in cmb]

move_block_down = partial(move_block, movement=np.array((0, -1)))

def check_overlap(cmb):
    if min(cmb,key=lambda item:item[0])[0] < 0 or max(cmb,key=lambda item:item[0])[0] >= 7 or min(cmb,key=lambda item:item[1])[1] < 0:
        return True
    if any([tuple(x) in settled for x in cmb]):
        return True
    return False

def remove_from_set(d, height):
    max_height = max(d,key=lambda item:item[1])[1]
    return(set([x for x in d if x[1] >= (max_height-height)]))
```
