```python
import networkx as nx
import numpy as np
targets = {"E": "G", "G": "E"}
G = nx.Graph()
cartCharDict = {'<': "W",'>': "E",'v': "S",'^': "N"}
cartList = []
for y, r in enumerate(data):
    for x, char in enumerate(r):
        if char in [".","G","E"]:
            if char == ".":
                G.add_node((x,y), unit=char, hp=0, ap=0)
            else:
                G.add_node((x,y), unit=char, hp=200, ap=3)
G.nodes(data=True)   

def createEdges(start):
    G.remove_edges_from(G.edges)
    for n,data in G.nodes(data=True):
        if data['unit'] == '.' or n == start:
            for neigh in getNeighbours(n):
                if neigh in G.nodes():
                    if G.nodes[neigh]['unit'] == ".":
                        G.add_edge(n,neigh)
def createOrder():
    orderList = []
    for n,data in G.nodes(data=True):
        if data['unit'] in ["G","E"]:
            orderList.append([n,data])
    return sorted(orderList, key=lambda x: (x[0][1], x[0][0]))

def move(unit):
    n_unit, data_self = unit
    createEdges(n_unit)
    in_range = set()
    for n_other,data_other in G.nodes(data=True):
        if n_unit == n_other or data_other['unit'] == "." or not data_other['unit'] == targets[data_self['unit']]:
            continue
        for neigh in getNeighbours(n_other):
            if neigh in G.nodes():
                if G.nodes[neigh]['unit'] == ".":
                    in_range.add(neigh)
    #print(f"{n_unit} in_range {in_range}")
    reachable = set([x for x in in_range if x in nx.node_connected_component(G, n_unit)])
    if not reachable:
        #print("not reachable")
        return n_unit, G.nodes[n_unit]['unit'],G.nodes[n_unit]['ap']
    #print(f"{n_unit} reachable {reachable}")
    #shortest_distance = min([manhattan_distance(n_unit,x) for x in reachable])
    shortest_distance = min([nx.shortest_path_length(G, source=n_unit, target=x) for x in reachable])
    nearest = [x for x in reachable if nx.shortest_path_length(G, source=n_unit, target=x) == shortest_distance]
    #print(f"{n_unit} nearest {nearest}")
    chosen = sorted(nearest, key=lambda x: (x[1], x[0]))[0]
    #print(f"{n_unit} chosen {chosen}")
    shortest_paths = list(nx.all_shortest_paths(G, source=n_unit, target=chosen))
    #print(f"shortest paths {shortest_paths}")
    possible_steps = set([x[1] for x in shortest_paths])
    #print(f"possible steps {possible_steps}")
    step = sorted(list(possible_steps), key=lambda x: (x[1], x[0]))[0]
    #print(f"chosen step {step}")
    #print("\n")
    G.nodes[step]['unit'] = data_self['unit']
    G.nodes[step]['hp'] = data_self['hp']
    G.nodes[step]['ap'] = data_self['ap']
    G.nodes[n_unit]['unit'] = "."
    G.nodes[n_unit]['hp'] = 0
    G.nodes[n_unit]['ap'] = 0
    return step, G.nodes[step]['unit'],G.nodes[step]['ap']
#def attack(unit):
    
    
def manhattan_distance(a, b):
    return np.abs(np.array(a) - np.array(b)).sum()
        
#def drawGraph():
board2str()    
```

```python
#createEdges((1,1))
#G.edges()
#nx.draw_kamada_kawai(G)
#order = createOrder()
num_rounds = 0
running = True
print(f"initial state\n----------")
board2str() 
print("----------")
while running:
    order = createOrder()
    deletedSet = set()
    for unit in order:
        if not running:
            continue
        if len([x for x in G.nodes(data=True) if x[1]['unit'] == 'G']) == 0 or len([x for x in G.nodes(data=True) if x[1]['unit'] == 'E']) == 0:
            print(f"all monsters dead {num_rounds}")
            board2str() 
            total_hp = sum([x[1]['hp'] for x in G.nodes(data=True)]) #+ 3 #hack
            print(f"total hp left {total_hp}")
            print(f"solution {num_rounds * total_hp}")
            #print(f"solution {(num_rounds-1) * total_hp}")
            #print(f"solution {num_rounds * (total_hp-3)}")
            #print(f"solution {(num_rounds-1) * (total_hp-3)}")
            running = False
            continue
        #print(unit[0])
        if unit[0] in deletedSet:
            print("skipping turn, because dead")
            continue
        #print(f"1: {unit[0],unit[1]['unit']}")
        neighbours = [[y,G.nodes[y]["hp"]] for y in [x for x in getNeighbours(unit[0]) if x in G.nodes()] if G.nodes[y]['unit'] == targets[unit[1]['unit']]]
        m_ap = unit[1]['ap']
        if len(neighbours) == 0:
            new_pos, m_type, m_ap = move(unit)
            neighbours = [[y,G.nodes[y]["hp"]] for y in [x for x in getNeighbours(new_pos) if x in G.nodes()] if G.nodes[y]['unit'] == targets[m_type]]
            #print(f"2: {new_pos,m_type}")
        #print(neighbours)
        if len(neighbours) > 0:
            target = sorted(neighbours, key=lambda x: (x[1], x[0][1], x[0][0]))[0]
            G.nodes[target[0]]['hp'] -= m_ap
            if G.nodes[target[0]]['hp'] <= 0:
                print("monster killed")
                G.nodes[target[0]]['unit'] = '.'
                G.nodes[target[0]]['hp'] = 0
                G.nodes[target[0]]['ap'] = 0
                #G.nodes[target[0]] = {'unit': '.', 'hp': 0, 'ap': 0}
                deletedSet.add(target[0])
            #print("already neighbour, attack")
    if not running:
        continue
    num_rounds += 1
    #print(f"round {num_rounds} results\n----------")
    #board2str() 
    #print("----------")
```
