```python
import copy
data = """
Before: [3, 2, 1, 1]
9 2 1 2
After:  [3, 2, 2, 1]
"""
with open("input_16.txt", 'r') as f:
    data = f.read()
    data = data.split("\n\n\n\n")
    machine = data[1]
    data = data[0]

data = [x for x in data.split("\n\n") if not x == ""]
data = [x.split("\n") for x in data]
for i, row in enumerate(data):
    data[i] = [x for x in data[i] if not x == ""]
print(data)
#data= data.replace("\n", "")
```

```python
import ast
matchesThree = 0
for row in data:
    numMatch = 0 
    registers = ast.literal_eval(row[0].split(": ")[1])
    instruction = [int(x) for x in row[1].split(" ")]
    registers_after = ast.literal_eval(row[2].split(":  ")[1])
    opp = instruction[0]
    valA = instruction[1]
    regA = registers[valA]
    valB = instruction[2]
    regB = registers[valB]
    regC = instruction[3]
    #addr
    registers_C = copy.deepcopy(registers)
    registers_C[regC] = regA + regB
    if registers_C == registers_after:
        numMatch += 1
    #addi
    registers_C = copy.deepcopy(registers)
    registers_C[regC] = regA + valB
    if registers_C == registers_after:
        numMatch += 1
    #mulr
    registers_C = copy.deepcopy(registers)
    registers_C[regC] = regA * regB
    if registers_C == registers_after:
        numMatch += 1
    #muli
    registers_C = copy.deepcopy(registers)
    registers_C[regC] = regA * valB
    if registers_C == registers_after:
        numMatch += 1
    #banr
    registers_C = copy.deepcopy(registers)
    registers_C[regC] = regA & regB
    if registers_C == registers_after:
        numMatch += 1
    #bani
    registers_C = copy.deepcopy(registers)
    registers_C[regC] = regA & valB
    if registers_C == registers_after:
        numMatch += 1
    #borr
    registers_C = copy.deepcopy(registers)
    registers_C[regC] = regA | regB
    if registers_C == registers_after:
        numMatch += 1
    #bori
    registers_C = copy.deepcopy(registers)
    registers_C[regC] = regA | valB
    if registers_C == registers_after:
        numMatch += 1
    #setr
    registers_C = copy.deepcopy(registers)
    registers_C[regC] = regA 
    if registers_C == registers_after:
        numMatch += 1
    #seti
    registers_C = copy.deepcopy(registers)
    registers_C[regC] = valA 
    if registers_C == registers_after:
        numMatch += 1
    #gtir
    registers_C = copy.deepcopy(registers)
    registers_C[regC] = int(valA > regB)
    if registers_C == registers_after:
        numMatch += 1
    #gtri
    registers_C = copy.deepcopy(registers)
    registers_C[regC] = int(regA > valB)
    if registers_C == registers_after:
        numMatch += 1
    #gtrr
    registers_C = copy.deepcopy(registers)
    registers_C[regC] = int(regA > regB)
    if registers_C == registers_after:
        numMatch += 1
    #eqir
    registers_C = copy.deepcopy(registers)
    registers_C[regC] = int(valA == regB)
    if registers_C == registers_after:
        numMatch += 1
    #eqri
    registers_C = copy.deepcopy(registers)
    registers_C[regC] = int(regA == valB)
    if registers_C == registers_after:
        numMatch += 1
    #eqrr
    registers_C = copy.deepcopy(registers)
    registers_C[regC] = int(regA == regB)
    if registers_C == registers_after:
        numMatch += 1 
    if numMatch >= 3:
        #print(numMatch, registers, registers_after)
        matchesThree += 1
print(matchesThree)
```
