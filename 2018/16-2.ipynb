```python
import copy
data = """
Before: [3, 2, 1, 1]
9 2 1 2
After:  [3, 2, 2, 1]
"""
with open("input_16.txt", 'r') as f:
    data = f.read()
    data = data.split("\n\n\n\n")
    machine = data[1]
    data = data[0]

data = [x for x in data.split("\n\n") if not x == ""]
data = [x.split("\n") for x in data]
for i, row in enumerate(data):
    data[i] = [x for x in data[i] if not x == ""]
#print(data)
#data= data.replace("\n", "")
opcodes = {}
for x in range(16):
    opcodes[x] = set()
```

```python
import ast
foundSet = set()
found = []
t = 0
while not t == 1:
    t += 1
    for row in data:
        numMatch = 0 
        registers = ast.literal_eval(row[0].split(": ")[1])
        instruction = [int(x) for x in row[1].split(" ")]
        registers_after = ast.literal_eval(row[2].split(":  ")[1])
        opp = instruction[0]
        valA = instruction[1]
        regA = registers[valA]
        valB = instruction[2]
        regB = registers[valB]
        regC = instruction[3]
        #addr
        registers_C = copy.deepcopy(registers)
        registers_C[regC] = regA + regB
        if registers_C == registers_after and not opp in foundSet:
            opcodes[opp].add("addr")
            found = [opp,"addr"]
            numMatch += 1
        #addi
        registers_C = copy.deepcopy(registers)
        registers_C[regC] = regA + valB
        if registers_C == registers_after and not opp in foundSet:
            numMatch += 1
            opcodes[opp].add("addi")
            found = [opp,"addi"]
        #mulr
        registers_C = copy.deepcopy(registers)
        registers_C[regC] = regA * regB
        if registers_C == registers_after and not opp in foundSet:
            numMatch += 1
            opcodes[opp].add("mulr")
        #muli
        registers_C = copy.deepcopy(registers)
        registers_C[regC] = regA * valB
        if registers_C == registers_after and not opp in foundSet:
            numMatch += 1
            opcodes[opp].add("muli")
        #banr
        registers_C = copy.deepcopy(registers)
        registers_C[regC] = regA & regB
        if registers_C == registers_after and not opp in foundSet:
            numMatch += 1
            opcodes[opp].add("banr")
        #bani
        registers_C = copy.deepcopy(registers)
        registers_C[regC] = regA & valB
        if registers_C == registers_after and not opp in foundSet:
            numMatch += 1
            opcodes[opp].add("bani")
        #borr
        registers_C = copy.deepcopy(registers)
        registers_C[regC] = regA | regB
        if registers_C == registers_after and not opp in foundSet:
            numMatch += 1
            opcodes[opp].add("borr")
        #bori
        registers_C = copy.deepcopy(registers)
        registers_C[regC] = regA | valB
        if registers_C == registers_after and not opp in foundSet:
            numMatch += 1
            opcodes[opp].add("bori")
        #setr
        registers_C = copy.deepcopy(registers)
        registers_C[regC] = regA 
        if registers_C == registers_after and not opp in foundSet:
            numMatch += 1
            opcodes[opp].add("setr")
        #seti
        registers_C = copy.deepcopy(registers)
        registers_C[regC] = valA 
        if registers_C == registers_after and not opp in foundSet:
            numMatch += 1
            opcodes[opp].add("seti")
        #gtir
        registers_C = copy.deepcopy(registers)
        registers_C[regC] = int(valA > regB)
        if registers_C == registers_after and not opp in foundSet:
            numMatch += 1
            opcodes[opp].add("gtir")
        #gtri
        registers_C = copy.deepcopy(registers)
        registers_C[regC] = int(regA > valB)
        if registers_C == registers_after and not opp in foundSet:
            numMatch += 1
            opcodes[opp].add("gtri")
        #gtrr
        registers_C = copy.deepcopy(registers)
        registers_C[regC] = int(regA > regB)
        if registers_C == registers_after and not opp in foundSet:
            numMatch += 1
            opcodes[opp].add("gtrr")
        #eqir
        registers_C = copy.deepcopy(registers)
        registers_C[regC] = int(valA == regB)
        if registers_C == registers_after and not opp in foundSet:
            numMatch += 1
            opcodes[opp].add("eqir")
        #eqri
        registers_C = copy.deepcopy(registers)
        registers_C[regC] = int(regA == valB)
        if registers_C == registers_after and not opp in foundSet:
            numMatch += 1
            opcodes[opp].add("eqri")
        #eqrr
        registers_C = copy.deepcopy(registers)
        registers_C[regC] = int(regA == regB)
        if registers_C == registers_after and not opp in foundSet:
            numMatch += 1 
            opcodes[opp].add("eqrr")
        #if numMatch == 1:
        #    print(opp)
        #    print(numMatch, registers, registers_after, found[0],found[1])
        #    opcodes[found[0]] = found[1]
        #    foundSet.add(found[0])
```
