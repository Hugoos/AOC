```python
class computer:
    def __init__(self, data, new_a):
        registers_s, program_s = data.split("\n\n")
        registers_s = registers_s.split("\n")
        self.registers = [0,0,0]
        for i in range(3):
            self.registers[i] = int(registers_s[i].split(": ")[1])
        self.registers[0] = new_a
        self.program = [int(x) for x in program_s.split(": ")[1].split(",")]
        self.ptr = 0
        self.output = []
        
    def outputIsProgram(self):
        return self.output == self.program
        
    def reset(self, data, new_a):
        registers_s, program_s = data.split("\n\n")
        registers_s = registers_s.split("\n")
        self.registers = [0,0,0]
        for i in range(3):
            self.registers[i] = int(registers_s[i].split(": ")[1])
        self.registers[0] = new_a
        self.program = [int(x) for x in program_s.split(": ")[1].split(",")]
        self.ptr = 0
        self.output = []
        
    def combo(self, op):
        if 0 <= op <= 3:
            return op
        if 4 <= op <= 6:
            return self.registers[op-4]
        if op == 7:
            print("op 7 found")
        else:
            print("error operand out of bounds")
        
    def adv(self, op): #0
        self.registers[0] = int(self.registers[0] // (2**self.combo(op)))
        
    def bxl(self, op): #1
        self.registers[1] = int(self.registers[1] ^ op)
    
    def bst(self, op): #2
        self.registers[1] = int(self.combo(op) % 8)
        
    def jnz(self, op): #3
        if not self.registers[0] == 0:
            self.ptr = op
            return True
        return False    
    
    def bxc(self, op): #4
        self.registers[1] = int(self.registers[1] ^ self.registers[2])
    
    def out(self,op): #5
        self.output.append(int(self.combo(op) % 8))
        
    def bdv(self, op): #6
        self.registers[1] = int(self.registers[0] // (2**self.combo(op)))
        
    def cdv(self, op): #7
        self.registers[2] = int(self.registers[0] // (2**self.combo(op)))
        
    def run(self):
        while self.ptr < len(self.program):
            instruction, op = self.program[self.ptr], self.program[self.ptr+1]
            self.ptr += 2
            if instruction == 0:
                self.adv(op)
            elif instruction == 1:
                self.bxl(op)
            elif instruction == 2:
                self.bst(op)
            elif instruction == 3:
                self.jnz(op)
            elif instruction == 4:
                self.bxc(op)
            elif instruction == 5:
                self.out(op)
            elif instruction == 6:
                self.bdv(op)
            elif instruction == 7:
                self.cdv(op)
            else:
                print("error: instruction out of bounds")
    def __str__(self):
        return f"register A: {self.registers[0]}\nregister B: {self.registers[1]}\nregister C: {self.registers[2]}\n\nprogram: {self.program}\npointer: {self.ptr}\noutput: {','.join(map(str, self.output))}\n"

c = computer(data, 0)
A_i = 0
while A_i < 8**3:
    if A_i % 1000 == 0:
        print(A_i)
    c.reset(data,A_i)
    c.run()
    print(A_i, c.output)
    if c.outputIsProgram():
        print(A_i)
        break
    A_i += 1
```
