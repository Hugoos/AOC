```python
import re
import copy
from helpers import npNicePrint as nprint
data = """
value 5 goes to bot 2
bot 2 gives low to bot 1 and high to bot 0
value 3 goes to bot 1
bot 1 gives low to output 1 and high to bot 0
bot 0 gives low to output 2 and high to output 0
value 2 goes to bot 2
"""
with open("input_10.txt", 'r') as f:
    data = f.read()

#
print(len(data))
#data = "".join([x for x in data if not x == ""])
data = data.lstrip('\n')
data = data.rstrip('\n')
data = [y.split(" ") for y in [x for x in data.split("\n") if not x == ""]]
print(len(data))
data
```

```python
bots = {}
outputs = {}
for row in data:
    if row[0] == "value":
        if not int(row[-1]) in bots:
            bots[int(row[-1])] = [[],-1,-1]
        bots[int(row[-1])][0].append(int(row[1]))
    elif row[0] == "bot":
        bot = int(row[1])
        type_ = row[5]
        val = int(row[6])
        type2_ = row[-2]
        val2 = int(row[-1])
        if not bot in bots:
            bots[bot] = [[],-1,-1]
        if type_ == "bot" and not val in bots:
            bots[val] = [[],-1,-1]
        if type2_ == "bot" and not val2 in bots:
            bots[val2] = [[],-1,-1]
        if type_ == "output" and not val in outputs:
            outputs[val] = []
        if type2_ == "output" and not val2 in outputs:
            outputs[val2] = []
        if type_ == "bot":
            bots[bot][1] = val
        if type2_ == "bot":
            bots[bot][2] = val2
        if type_ == "output":
            bots[bot][1] = f"o{val}"
        if type2_ =="output":
            bots[bot][2] = f"o{val2}"
        
print(bots)
print(outputs)
```

```python
comparing = [17,61]
quit = False
while not quit:
    print(bots)
    print(outputs)
    bots_n = copy.deepcopy(bots)
    for k,v in bots.items():
        if len(v[0]) >= 2:
            arr = v[0]
            arr.sort()
            if arr == comparing:
                print(f"solution: {k}")
                quit = True
            low_target = v[1]
            high_target = v[2]
            if "o" in str(low_target):
                t = outputs[int(low_target[1:])]
                t.append(arr.pop(0))
                outputs[int(low_target[1:])] = t
            else:
                t = bots_n[int(low_target)][0]
                t.append(arr.pop(0))
                bots_n[int(low_target)][0] = t
            if "o" in str(high_target):
                t = outputs[int(high_target[1:])]
                t.append(arr.pop(0))
                outputs[int(high_target[1:])] = t
            else:
                t = bots_n[int(high_target)][0]
                t.append(arr.pop(0))
                bots_n[int(high_target)][0] = t
            bots_n[k][0] = []
    bots = bots_n
            
        
```
