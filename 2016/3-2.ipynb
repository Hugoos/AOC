```python
import copy
import re
data = """
101 301 501
102 302 502
103 303 503
201 401 601
202 402 602
203 403 603
"""
with open("input_3.txt", 'r') as f:
    data = f.read()
print(data)
data = [re.sub(' +', ' ', x.lstrip(' ')) for x in data.split("\n") if not x == ""]
print(data)
data = [int(data[z].split(" ")[j]) for j in [i for i in range(3)] for z in range(len(data)) if not data[z].split(" ")[j] == ""]
data = [data[i:i+3] for i in range(0, len(data), 3)]
#data = ["ieodomkazucvgmuy"]
print(data)
```

```python
import itertools as it
num_triangle = 0
for triangle in data:
    print(triangle)
    is_triangle = True
    for combination in list(it.combinations(triangle,2)):
        triangle_c = copy.deepcopy(triangle)
        triangle_c.remove(combination[0])
        triangle_c.remove(combination[1])
        if combination[0] + combination[1] <= triangle_c[0]:
            is_triangle = False
    if is_triangle:
        num_triangle += 1
print(num_triangle)
        
            
```
