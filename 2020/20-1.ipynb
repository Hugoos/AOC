```python
tiles = test.split("\n\n")
```

```python
import numpy as np
def toNumpy(r):
    row = np.zeros(len(r))
    convertDict = {".": 0, "#": 1}
    for i, cell in enumerate(r):
        row[i] = convertDict[cell]
    return row
```

```python
tilesDict = {}
for tile in tiles:
    rows = [ x for x in tile.split("\n") if not x == ""]
    if not rows:
        continue
    tile_np = np.zeros((len(rows[1:]), len(rows[1])))
    
    tid = int(rows[0].split(" ")[1][:-1])
    for i, r in enumerate(rows[1:]):
        tile_np[i] = toNumpy(r)
    tilesDict[tid] = tile_np
```

```python
def getAllConfigurations(arr):
    configurationList = [arr]
    configurationList.append(np.fliplr(arr))
    for i in range(3):
        arr = np.rot90(arr)
        configurationList.append(arr)
        configurationList.append(np.fliplr(arr))
    return configurationList

def checkLineUp(arr1, arr2):
    con1 = getAllConfigurations(arr1)
    con2 = getAllConfigurations(arr2)
    for grid1 in con1:
        for grid2 in con2:
            if np.array_equal(grid1[0], grid2[-1]):
                #print(f"arr1 connects arr2 top to bottom")
                #npNicePrint(grid1)
                #npNicePrint(grid2)
                return True
            if np.array_equal(grid1[-1], grid2[0]):
                #print(f"arr1 connects arr2 bottom to top")
                #npNicePrint(grid1)
                #npNicePrint(grid2)
                return True
            if np.array_equal(grid1[:,0], grid2[:,-1]):
                #print(f"arr1 connects arr2 right to left")
                #npNicePrint(grid1)
                #npNicePrint(grid2)
                return True
            if np.array_equal(grid1[:,-1], grid2[:,0]):
                #print(f"arr1 connects arr2 left to right")
                #npNicePrint(grid1)
                #npNicePrint(grid2)
                return True
    return False
    
```

```python
checkLineUp(tilesDict[1951], tilesDict[2473])
```

```python
import networkx as nx
G = nx.Graph()
G.add_nodes_from(tilesDict.keys())
print(G.nodes)
for key in tilesDict.keys():
    for key2 in tilesDict.keys():
        if not key == key2:
            if checkLineUp(tilesDict[key], tilesDict[key2]):
                G.add_edge(key,key2)
```

```python
def npNicePrint(arr):
    convertDict = {0: ".", 1: "#"}
    s = ""
    for r in arr:
        for c in r:
            s += convertDict[c]
        s += "\n"
    print(s)
        
```
