```python
import ast
d = {}

test, codes = test.split("\n\n")

codes = [ x for x in codes.split("\n") if not x == ""]

for line in test.split("\n"):
    if not ":" in line:
        break
    num, line = line.split(":")
    num = int(num)
    d[num] = []
    if "|" in line:    
        part1, part2 = line.split("|")
        part1 = [int(x) for x in part1.split(" ") if not x == ""]
        part2 = [int(x) for x in part2.split(" ") if not x == ""]
        d[num].append(part1)
        d[num].append(part2)
    else:
        part1 = [ast.literal_eval(x) for x in line.split(" ") if not x == ""]
        d[num] = [part1]
```

```python
def explore(val, depth = 0):
    #if depth == 6:
    #    return [""]
    ansList = [] 
    for x in d[val]:
        ans = [""]
        for y in x:
            if isinstance(y, str):
                return y
            addThing = explore(y, depth+1)
            #print(ans, addThing)
            ans = [q + z for q in ans for z in addThing]
            #print(ans, "\n")
            #print(ans)
        ansList.extend(ans)
    return ansList
```

```python
solutions42 = explore(42)
solutions31 = explore(31)
```

```python
solutions = explore(0)
solutions = [x for x in codes if x in solutions]
```
