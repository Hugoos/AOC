```python
test = """
mxmxvkd kfcds sqjhc nhms (contains dairy, fish)
trh fvjkl sbzzf mxmxvkd (contains dairy)
sqjhc fvjkl (contains soy)
sqjhc mxmxvkd sbzzf (contains fish)
"""
with open("input_21.txt", 'r') as f:
    test = f.read()
```

```python
rows = [ x for x in test.split("\n") if not x == ""]
ingredientsDict = {}
ingredientsList = []
for r in rows:
    ingredients, allergens = [ x for x in r.split(" (") if not x == ""]
    ingredients = [x.replace(" ", "") for x in ingredients.split(" ")]
    allergens = allergens.replace("contains ", "")
    allergens = allergens.replace(")", "")
    allergens = [x.replace(" ", "") for x in allergens.split(",")]
    for ing in ingredients:
        if not ing in ingredientsDict.keys():
            ingredientsDict[ing] = 0
        #ingredientsDict[ing].append(allergens)
        ingredientsDict[ing] += 1
    ingredientsList.append([set(ingredients), allergens])
```

```python
def getAllSingle(l):
    singleList = []
    for food in l:
        add = True
        ingredients = food[0]
        allergens = food[1]
        if len(allergens) == 1:
            for food2 in l:
                if not food == food2:
                    if allergens[0] in food2[1]:
                        add = False
            if add:
                singleList.append(food)
    return(singleList)

def getAllCommon(l, food):
    commonList = []
    ingredients = food[0]
    allergens = food[1]
    for food2 in l:
        add = False
        if not food == food2:
            for allergen in allergens:
                if allergen in food2[1]:
                    add = True
        if add:
            commonList.append(food2)
    return commonList
```

```python
ingredientsList2 = []
if getAllSingle(ingredientsList):
    for food in getAllSingle(ingredientsList):
        ingr = food[0]
        for food2 in ingredientsList:
            food2[0] = food2[0] - ingr
            if not len(food2[0]) == 0:
                ingredientsList2.append(food2)
    ingredientsList = ingredientsList2 
```

```python
reducedIngredientsList = set()
for food in ingredientsList:
    ingredients = food[0]
    for food2 in getAllCommon(ingredientsList,food):
        ingredients = ingredients - food2[0]
    reducedIngredientsList = reducedIngredientsList | ingredients
```
